# レポート（応用数学）

## 第1章 : 線形代数
- 名称
    - 「線形代数（学）」は「linear algebra」の邦訳である
    - 「線形代数」と「線型代数」の2通りの表記がある
        - 空間や写像の「形」でなく「型」という理由等で敢えて後者の表記が採用される場面もある
- スカラー・ベクトル・行列
    - スカラー
        - 階数$0$のテンソルに対応
        - いわゆる普通の数
            - 例.
        $$
        a=1
        $$
            - 大抵は実数$\mathbb{R}$
        - 四則演算可能
        - ベクトルに対する係数になれる
    - ベクトル
        - 階数$1$のテンソルに対応
        - 「大きさ」と「向き」を持ち、幾何学的座標系に於いて矢印（有向線分）で図示される
        - スカラーとセットで表示（成分表示）され、スカラーの要素を水平方向（横）に並べたものを「行ベクトル」、垂直方向（縦）に並べたものを「列ベクトル」という
        $$
        \left(
        \begin{array}{ccc}
            2 & 0 & -1
        \end{array}
        \right),\hspace{5pt}
        \left(
        \begin{array}{c}
            2 \\
            0 \\
            -1
        \end{array}
        \right)
        $$
        - ベクトルの文字表記は一般には$\rightarrow$記号を冠した小文字（例: $\vec{a}$）又は太字の小文字（例: $\mathbf{a}$）が使用される
        - 列ベクトル$\vec{a}$に紐付けられた$n$個のスカラーの要素を$\vec{a}$の「成分」といい、各成分固有の添字（インデックス）が割り当てられた列ベクトル
        $$
        \vec{a}=\left(
        \begin{array}{c}
            a_{1} \\
            a_{2} \\
            \vdots \\
            a_{n}
        \end{array}
        \right)
        $$
        を「$n$次元列ベクトル」という
            - $1$から$n$迄の範囲内の任意の成分を表わす添字には$i$を用いる
            $$
            1 \leqq i \leqq n \hspace{10pt} (i, n \in \mathbb{N})
            $$
            - $a_{i}$を$\vec{a}$の「第$i$成分」という
        - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
        
        ```python
        import numpy as np
        
        # 列ベクトルのサイズ
        n = 3
        # 成分の値が一様乱数のn次元列ベクトル
        vec_a = np.random.rand(n)
        #vec_a = np.random.rand(n).reshape(-1, 1)
        
        print(vec_a)
        ```
        
        - 演算
            - スカラーとベクトルの積（スカラー乗法）
        $$
        \begin{eqnarray}
            c\vec{a}&=&\left(
            \begin{array}{c}
                ca_{1} \\
                ca_{2} \\
                \vdots \\
                ca_{n}
            \end{array}
            \right) \\
            2\left(
            \begin{array}{c}
                2 \\
                -3 \\
                1
            \end{array}
            \right)
            &=&\left(
            \begin{array}{c}
                2\cdot 2 \\
                2\cdot (-3) \\
                2\cdot 1
            \end{array}
            \right)
            =\left(
            \begin{array}{c}
                4 \\
                -6 \\
                2
            \end{array}
            \right)
        \end{eqnarray}
        $$
            - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
            
            ```python
            import numpy as np

            # スカラーc
            c = 2
            # ベクトルa
            vec_a = np.array([2, -3, 1], dtype=int)

            print(c * vec_a)
            ```
            
    - 行列
        - 階数$2$のテンソルに対応
        - スカラーを表にしたもの
        $$
        \left(
        \begin{array}{cccc}
            1 & 3 & 0 & -1 \\
            2 & 5 & \frac{1}{2} & 6 \\
            -2 & 3 & -10 & \frac{1}{4} 
        \end{array}
        \right)
        $$
        - 行列の文字表記は一般には太字の大文字（例: $\mathbf{A}$）が使用される
        - 行列$\mathbf{A}$に紐付けられたスカラーの要素を$\mathbf{A}$の「成分」といい、各成分固有の添字が行と列の各軸毎に割り当てられた行列
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{m1} & a_{m2} & \ldots & a_{mn}
        \end{array}
        \right)
        $$
        を「$m$行$n$列の行列」又は「$(m,n)$型の行列」という
            
            - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
            
            ```python
            import numpy as np
            
            # 行列の行数と列数
            m, n = 3, 4
            # 成分の値が一様乱数の(m,n)型の行列
            A = np.random.rand(m, n)
            
            print(A)
            ```
            
            - $1$から$m$迄の範囲内の任意の$n$次元行ベクトル（タプル）を表わす添字には$i$を用いる
            $$
            1 \leqq i \leqq m \hspace{10pt} (i, m \in \mathbb{N})
            $$
            - 行ベクトル$a_{i}$を$\mathbf{A}$の「第$i$行」という
            - $1$から$n$迄の範囲内の任意の$m$次元列ベクトル（カラム）を表わす添字には$j$を用いる
            $$
            1 \leqq j \leqq n \hspace{10pt} (j, n \in \mathbb{N})
            $$
            - 列ベクトル$a_{j}$を$\mathbf{A}$の「第$j$列」という
            - $a_{ij}$を$\mathbf{A}$の「$(i,j)$成分」という
            - $\mathbf{A}$の$(i,j)$成分を$(j,i)$成分の位置に移動して作った新たな行列
        $$
        \begin{eqnarray}
            \mathbf{A}^{\mathrm{T}}&=&\left(
            \begin{array}{cccc}
                a_{11} & a_{12} & \ldots & a_{1n} \\
                a_{21} & a_{22} & \ldots & a_{2n} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{m1} & a_{m2} & \ldots & a_{mn}
            \end{array}
            \right)^{\mathrm{T}}&=&\left(
            \begin{array}{cccc}
                a_{11} & a_{21} & \ldots & a_{m1} \\
                a_{12} & a_{22} & \ldots & a_{m2} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{1n} & a_{2n} & \ldots & a_{mn}
            \end{array}
            \right) \\
            &=&\left(
            \begin{array}{cccc}
                1 & 3 & 0 & -1 \\
                2 & 5 & \frac{1}{2} & 6 \\
                -2 & 3 & -10 & \frac{1}{4} 
            \end{array}
            \right)^{\mathrm{T}}
            &=&\left(
            \begin{array}{ccc}
                1 & 2 & -2 \\
                3 & 5 &  3 \\
                0 & \frac{1}{2} & -10 \\
                -1 & 6 & \frac{1}{4}
            \end{array}
            \right)
        \end{eqnarray}
        $$
            を$\mathbf{A}$の「転置行列」という
                
                - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>

                ```python
                import numpy as np

                # 行列の行数と列数
                m, n = 3, 4
                # 成分の値が一様乱数の(m,n)型の行列
                A = np.random.rand(m, n)
                # Aの転置行列
                tA = A.T

                print(A)
                print(tA)
                ```

            - 行数と列数が等しい$(m=n)$行列
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{n1} & a_{n2} & \ldots & a_{nn}
        \end{array}
        \right)
        \ni \left(
        \begin{array}{ccc}
            1 & 3 & 0 \\
            2 & 5 & \frac{1}{2} \\
            -2 & 3 & -10
        \end{array}
        \right)
        $$
            を「正方行列」といい、$n$行$n$列の正方行列を「$n$次正方行列」又は「$n$次行列」という
                
                - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
                
                ```python
                import numpy as np
                
                # 正方行列のサイズ
                n = 3
                # 成分の値が一様乱数のn次正方行列
                A = np.random.rand(n, n)
                
                print(A)
                ```

            - $a_{11},a_{22},\ldots,a_{nn}$の成分を「対角成分」といい、対角成分を除く全ての成分の値を$0$とした正方行列
        $$
        \mathbf{C}=\left(
        \begin{array}{cccc}
            c_{1} & & & {\LARGE 0} \\
            & c_{2} & & \\
            & & \ddots & \\
            {\LARGE 0} & & & c_{n}
        \end{array}
        \right)
        \ni \left(
        \begin{array}{ccc}
            1 & 0 & 0 \\
            0 & 5 & 0 \\
            0 & 0 & -10
        \end{array}
        \right)
        $$
            を「対角行列」という
                
                - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
                
                ```python
                import numpy as np

                # 正方行列のサイズ
                n = 3
                # 成分の値が一様乱数のn次正方行列
                A = np.random.rand(n, n)
                # Aの対角成分で構成された対角行列
                C = np.diag(np.diag(v=A))

                print(C)
                ```
                
            - 全ての対角成分$c_{1},c_{2},\ldots,c_{n}$の値を$1$とした対角行列
        $$
        \mathbf{I}=\left(
        \begin{array}{cccc}
            1 & & & {\LARGE 0} \\
            & 1 & & \\
            & & \ddots & \\
            {\LARGE 0} & & & 1
        \end{array}
        \right)
        \ni \left(
        \begin{array}{ccc}
            1 & 0 & 0 \\
            0 & 1 & 0 \\
            0 & 0 & 1
        \end{array}
        \right)
        $$
            を「単位行列」という
                
                - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
                
                ```python
                import numpy as np
                
                # 正方行列のサイズ
                n = 3
                # 成分の値が一様乱数のn次正方行列
                A = np.random.rand(n, n)
                # Aと同じサイズの単位行列
                I = np.identity(n=A.shape[0])
                
                print(I)
                ```
        
        - 演算
            - 行列と列ベクトルの積
            <div style="overflow: auto hidden;">
            $$
            \begin{eqnarray}
                \mathbf{A}\vec{x}=
                \left(
                \begin{array}{cccc}
                    a_{11} & a_{12} & \ldots & a_{1n} \\
                    a_{21} & a_{22} & \ldots & a_{2n} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    a_{m1} & a_{m2} & \ldots & a_{mn}
                \end{array}
                \right)
                \left(
                \begin{array}{c}
                    x_{1} \\
                    x_{2} \\
                    \vdots \\
                    x_{n}
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{c}
                    \sum_{i=1}^{n}a_{1i}x_{i} \\
                    \sum_{i=1}^{n}a_{2i}x_{i} \\
                    \vdots \\
                    \sum_{i=1}^{n}a_{mi}x_{i} \\
                \end{array}
                \right) \\
                \left(
                \begin{array}{cc}
                    3 & 1 \\
                    2 & -3 \\
                    1 & 4
                \end{array}
                \right)
                \left(
                \begin{array}{c}
                    1 \\
                    3
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{c}
                    3 \cdot 1 + 1 \cdot 3 \\
                    2 \cdot 1 + (-3) \cdot 3 \\
                    1 \cdot 1 + 4 \cdot 3
                \end{array}
                \right) \\
                &=& 
                \left(
                \begin{array}{c}
                    6 \\
                    -7 \\
                    13
                \end{array}
                \right)
            \end{eqnarray}
            $$
            </div>
                
                - 行列$\mathbf{A}$の列数$n$と列ベクトル$\vec{x}$の次元数（成分の個数）$n$が等しいときにのみ積が成立する
            - 行列同士の積（内積、dot積）
            <div style="overflow: auto hidden;">
            $$
            \begin{eqnarray}
                \mathbf{A}\mathbf{B}=
                \left(
                \begin{array}{cccc}
                    a_{11} & a_{12} & \ldots & a_{1n} \\
                    a_{21} & a_{22} & \ldots & a_{2n} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    a_{m1} & a_{m2} & \ldots & a_{mn}
                \end{array}
                \right)
                \left(
                \begin{array}{cccc}
                    b_{11} & b_{12} & \ldots & b_{1n} \\
                    b_{21} & b_{22} & \ldots & b_{2n} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    b_{m1} & b_{m2} & \ldots & b_{mn}
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{cccc}
                    \sum_{i=1}^{n}a_{1i}b_{i1} & \sum_{i=1}^{n}a_{1i}b_{i2} & \ldots & \sum_{i=1}^{n}a_{1i}b_{in} \\
                    \sum_{i=1}^{n}a_{2i}b_{i1} & \sum_{i=1}^{n}a_{2i}b_{i2} & \ldots & \sum_{i=1}^{n}a_{2i}b_{in} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    \sum_{i=1}^{n}a_{mi}b_{i1} & \sum_{i=1}^{n}a_{mi}b_{i2} & \ldots & \sum_{i=1}^{n}a_{mi}b_{in}
                \end{array}
                \right) \\
                \left(
                \begin{array}{cc}
                    3 & 1 \\
                    2 & -3 \\
                    1 & 4
                \end{array}
                \right)
                \left(
                \begin{array}{ccc}
                    1 & -2 & 0 \\
                    3 &  1 & 2
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{ccc}
                    3 \cdot 1 + 1 \cdot 3    & 3 \cdot (-2) + 1 \cdot 1    & 3 \cdot 0 + 1 \cdot 2 \\
                    2 \cdot 1 + (-3) \cdot 3 & 2 \cdot (-2) + (-3) \cdot 1 & 2 \cdot 0 + (-3) \cdot 2 \\
                    1 \cdot 1 + 4 \cdot 3    & 1 \cdot (-2) + 4 \cdot 1    & 1 \cdot 0 + 4 \cdot 2
                \end{array}
                \right) \\
                &=& 
                \left(
                \begin{array}{ccc}
                     6 & -5 &  2 \\
                    -7 & -7 & -6 \\
                    13 &  2 &  8
                \end{array}
                \right)
            \end{eqnarray}
            $$
            </div>
                
                - 可換則（交換法則）は一般には成立しない
                - 行列$\mathbf{A}$の列数$n$と行列$\mathbf{B}$の行数$m$が等しいときにのみ積が成立する
                - 行列$\mathbf{A}$は行ベクトルを垂直方向に並べたもの、行列$\mathbf{B}$は列ベクトルを水平方向に並べたものとも考えられる

        - 主な用途
            - ベクトル等データの変換
                - データフロー図（DFD）でいう「データ処理」に相当
            - 連立方程式（連立一次方程式）を解く
                - $ax=b$の形で表したい
                - $n$本の一次方程式を持つ連立方程式
        $$
        \left\{
        \begin{array}{rrrrrrrrr}
            a_{11}x_{1} &+& a_{12}x_{2} &+& \ldots &+& a_{1n}x_{n} &=& b_{1} \\
            a_{21}x_{1} &+& a_{22}x_{2} &+& \ldots &+& a_{2n}x_{n} &=& b_{2} \\
            \vdots &\,& \vdots &\,& \ddots &\,& \vdots &\,& \vdots \\
            a_{n1}x_{1} &+& a_{n2}x_{2} &+& \ldots &+& a_{nn}x_{n} &=& b_{n}
        \end{array}
        \right.
        $$
                    を$x$の係数$a$を成分に持つ$n$次正方行列$\mathbf{A}$、$x$を成分に持つ$n$次元列ベクトル$\vec{x}$、$b$を成分に持つ$n$次元列ベクトル$\vec{b}$
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{n1} & a_{n2} & \ldots & a_{nn}
        \end{array}
        \right), 
        \vec{x}=\left(
        \begin{array}{c}
            x_{1} \\
            x_{2} \\
            \vdots \\
            x_{n}
        \end{array}
        \right), 
        \vec{b}=\left(
        \begin{array}{c}
            b_{1} \\
            b_{2} \\
            \vdots \\
            b_{n}
        \end{array}
        \right)
        $$
                    に当て嵌めたとき
        $$
        \mathbf{A}\vec{x}=\vec{b}
        $$
                    と表わすことができる。このとき、$\mathbf{A}$を「係数行列」という
                    - **連立方程式の行列の形は、連立方程式の係数を取り出したもの**といえる
                    - 例. 連立方程式
        $$
        \left\{
        \begin{array}{rrrrr}
            2x &-& 3y &=& 7 \\
            3x &+& 4y &=& 2
        \end{array}
        \right.
        $$
                    を
        $$
        \mathbf{A}=\left(
        \begin{array}{cc}
            2 & -3 \\
            3 & 4
        \end{array}
        \right),\hspace{5pt}
        \vec{x}=\left(
        \begin{array}{c}
            x \\
            y
        \end{array}
        \right),\hspace{5pt}
        \vec{b}=\left(
        \begin{array}{c}
            7 \\
            2
        \end{array}
        \right)
        $$
                    に当て嵌めたとき
        $$
        \begin{eqnarray}
            \mathbf{A}\vec{x}&=&\vec{b} \\
            \left(
            \begin{array}{cc}
                2 & -3 \\
                3 & 4
            \end{array}
            \right)
            \left(
            \begin{array}{c}
                x \\
                y
            \end{array}
            \right)
            &=&\left(
            \begin{array}{c}
                7 \\
                2
            \end{array}
            \right)
        \end{eqnarray}
        $$
                    と表わすことができる
                - 係数行列$\mathbf{A}$と、別の行列又はベクトル（例: $\vec{b}$）を水平方向に連結（concatenate）して作った新たな行列
        $$
        \left(
        \begin{array}{c|c}
            \mathbf{A} & \vec{b}
        \end{array}
        \right)
        =\left(
        \begin{array}{cc|c}
            2 & -3 & 7 \\
            3 & 4 & 2
        \end{array}
        \right)
        $$
                を「拡大係数行列」という
                    - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
                    
                    ```python
                    import numpy as np
                    
                    # 係数行列
                    A = np.array([2, -3, 3, 4], dtype=int).reshape(2, 2)
                    # 係数行列と水平方向に連結する為の列ベクトル
                    vec_b = np.array([7, 2], dtype=int).reshape(-1, 1)
                    # 拡大係数行列
                    ```

                    - 証明<br />
                    順列の異なる要素を交換する操作を「互換」といい、偶数回の互換で作れる順列を「偶順列（even permutation）」、奇数回の互換で作れる順列を「奇順列（odd permutation）」という。順列に対し、その偶奇に従って
                $$
                    \varepsilon(i, j, \ldots, k) =
                    \begin{cases}
                        +1 & ((i, j, \ldots, k)\,\text{is the even permutation}) \\
                        -1 & ((i, j, \ldots, k)\,\text{is the odd permutation})
                    \end{cases}
                $$
                    を定義し「順列$(i, j, \ldots, k)$の符号」という。
                    このように、順列の符号は互換によって変わるので
                $$
                    \varepsilon(i, \ldots, j, \ldots, k, \ldots, l)=-\varepsilon(i, \ldots, k, \ldots, j, \ldots, l)
                $$
                    を満たす。よって、$n$次の行列式
                $$
                    \mathrm{det}\mathbf{A}=\left|
                    \begin{array}{cccc}
                        a_{11} & a_{12} & \ldots & a_{1n} \\
                        a_{21} & a_{22} & \ldots & a_{2n} \\
                        \vdots & \vdots & \ddots & \vdots \\
                        a_{n1} & a_{n2} & \ldots & a_{nn}
                    \end{array}
                    \right|
                $$
                    に対し
                $$
                    \begin{eqnarray}
                        \text{LS} &=& \sum\varepsilon(i, \ldots, j, \ldots, k, \ldots, l)a_{1i}\ldots a_{rj}\ldots a_{sk} \ldots a_{nl} \\
                        &=& \sum\varepsilon(i, \ldots, j, \ldots, k, \ldots, l)a_{1i}\ldots a_{sk}\ldots a_{rj} \ldots a_{nl} \\
                        &=& \sum\{-\varepsilon(i, \ldots, k, \ldots, j, \ldots, l)a_{1i}\ldots a_{sk}\ldots a_{rj} \ldots a_{nl}\} \\
                        &=& \text{RS}
                    \end{eqnarray}
                $$
                    を満たすので、行列式の行を交換すると符号が変わる

    \right)
    $$
    に対し、$0$でない特殊なベクトル$\vec{x}$とスカラーの定数$\lambda$が存在して
    $$
    \mathbf{A}\vec{x}=\lambda\vec{x}
    $$
    を満たすとき、$\lambda$を$\mathbf{A}$の「固有値」、$\vec{x}$を$\lambda$に対する$\mathbf{A}$の「固有ベクトル」という
        - 例.
        $$
        \mathbf{A}=\left(
        \begin{array}{cc}
            4 & 2 \\
            -3 & -1
        \end{array}
        \right),\hspace{5pt}
        \vec{x}=\left(
        \begin{array}{c}
            2 \\
            -3
        \end{array}
        \right)
        $$
        のとき
        $$
        \mathbf{A}\vec{x}=\lambda\vec{x}
        $$
        $$
        \left(
        \begin{array}{cc}
            4 & 2 \\
            -3 & -1
        \end{array}
        \right)
        \left(
        \begin{array}{c}
            2 \\
            -3
        \end{array}
        \right)
        =\lambda\left(
        \begin{array}{c}
            2 \\
            -3
        \end{array}
        \right)
        $$
        より、$\lambda=1$のとき、$\left(\begin{array}{c}2 \\ -3\end{array}\right)$は、固有値$1$に対する$\left(\begin{array}{cc} 4 & 2 \\ -3 & -1 \end{array}\right)$の固有ベクトル（の定数倍）である
    - 固有値・固有ベクトルの求め方
    $$
    \begin{eqnarray}
        \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
        (\mathbf{A}-\lambda I)\vec{x}&=&\vec{0}
    \end{eqnarray}
    $$
    $\mathbf{A}-\lambda I$の逆行列$(\mathbf{A}-\lambda I)^{-1}$が存在しないこと、かつ$\vec{x} \neq 0$より
    $$
    |\mathbf{A}-\lambda I|=0
    $$
    を満たすとき、この方程式を「固有方程式」といい、その左辺を「固有多項式」という
        - 例.
    $$
    \mathbf{A}=\left(
    \begin{array}{cc}
        2 & 1 \\
        3 & 4
    \end{array}
    \right)
    $$
        のとき、固有方程式より
    $$
    \begin{eqnarray}
        |\mathbf{A}-\lambda I|&=&0 \\
        \left|
        \begin{array}{cc}
            2-\lambda & 1 \\
            3 & 4-\lambda
        \end{array}
        \right|&=&0 \\
        (2-\lambda)(4-\lambda)-1\cdot3&=&0 \\
        \lambda^{2}-6\lambda+5&=&0 \\
        (\lambda-1)(\lambda-5)&=&0 \\
        \lambda&=&5\,\mathrm{or}\,1
    \end{eqnarray}
    $$
        したがって、$\mathbf{A}$の固有値$\lambda$は$5$と$1$の2つである<br /><br />$\lambda=5$のとき
    $$
    \begin{eqnarray}
        \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
        \left(
        \begin{array}{cc}
            2 & 1 \\
            3 & 4
        \end{array}
        \right)
        \left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        &=&5\left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        \; \Longleftrightarrow \;
        \left\{
        \begin{array}{rrrrr}
            2x_{1} &+& x_{2} &=& 5x_{1} \\
            3x_{1} &+& 4x_{2} &=& 5x_{2}
        \end{array}
        \right.
    \end{eqnarray}
    $$
        よって
    $$
    3x_{1} = x_{2}
    $$
        したがって、固有値$\lambda_{1}=5$に対する$\mathbf{A}$の固有ベクトル$\vec{v}_{1}$は$\left(\begin{array}{c} 1 \\ 3 \end{array}\right)$の定数倍である<br />$\lambda=1$のとき
    $$
    \begin{eqnarray}
        \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
        \left(
        \begin{array}{cc}
            2 & 1 \\
            3 & 4
        \end{array}
        \right)
        \left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        &=&\left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        \; \Longleftrightarrow \;
        \left\{
        \begin{array}{rrrrr}
            2x_{1} &+& x_{2} &=& x_{1} \\
            3x_{1} &+& 4x_{2} &=& x_{2}
        \end{array}
        \right.
    \end{eqnarray}
    \\
    $$
        よって
    $$
    -x_{1} = x_{2}
    $$
        したがって、固有値$\lambda_{2}=1$に対する$\mathbf{A}$の固有ベクトル$\vec{v}_{2}$は$\left(\begin{array}{c} 1 \\ -1 \end{array}\right)$の定数倍である
    - 固有値分解（スペクトル分解）
        - 或る$n$次正方行列
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{n1} & a_{n2} & \ldots & a_{nn}
        \end{array}
        \right)
        $$
        が固有値$\lambda_{1},\lambda_{2},\ldots,\lambda_{n}$と固有ベクトル$\vec{v}_{1},\vec{v}_{2},\ldots,\vec{v}_{n}$を持つとき、各固有値の並びを対角成分とした対角行列
        $$
        \mathbf{\Lambda}=\left(
        \begin{array}{cccc}
            \lambda_{1} & & & {\LARGE 0} \\
            & \lambda_{2} & & \\
            & & \ddots & \\
            {\LARGE 0} & & & \lambda_{n}
        \end{array}
        \right)
        $$
        と、それに対する固有ベクトルを並べた行列
        $$
        \mathbf{V}=\left(
        \begin{array}{cccc}
            \vec{v}_{1} & \vec{v}_{2} & \ldots & \vec{v}_{n}
        \end{array}
        \right)
        $$
        を用意したとき、それらは
        $$
        \begin{eqnarray}
            \mathbf{A}\mathbf{V}&=&\mathbf{V}\mathbf{\Lambda} \\
            \mathbf{A}&=&\mathbf{V}\mathbf{\Lambda}\mathbf{V}^{-1}
        \end{eqnarray}
        $$
        を満たす
            - $\mathbf{\Lambda}$の成分$\lambda_{i}$の値の順序は基本的に降順
            - $\mathbf{V}$は定数倍なので一つに定まらない
                - 固有値分解に於いて、より重要なのは$\mathbf{\Lambda}$の方
            - 例
            $$
            \mathbf{A}=\left(
            \begin{array}{cc}
                2 & 1 \\
                3 & 4
            \end{array}
            \right),\hspace{5pt}
            \lambda_{1}=5,\hspace{5pt} \lambda_{2}=1,\hspace{5pt}
            \vec{v}_{1}=\left(\begin{array}{c} 1 \\ 3 \end{array}\right),\hspace{5pt}
            \vec{v}_{2}=\left(\begin{array}{c} 1 \\ -1 \end{array}\right)
            $$
            のとき
        $$
        \begin{eqnarray}
            \mathbf{A}&=&\mathbf{V}\mathbf{\Lambda}\mathbf{V}^{-1} \\
            \mathbf{A}
            &=&
            \left(
            \begin{array}{cc}
                \vec{v}_{1} &  \vec{v}_{2} \\
                & 
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                \lambda_{1} & 0 \\
                0           & \lambda_{2}
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                \vec{v}_{1} &  \vec{v}_{2} \\
                & 
            \end{array}
            \right)^{-1} \\
            \left(
            \begin{array}{cc}
                2 & 1 \\
                3 & 4
            \end{array}
            \right)
            &=&
            \left(
            \begin{array}{cc}
                1 &  1 \\
                3 & -1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                5 & 0 \\
                0 & 1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                1 &  1 \\
                3 & -1
            \end{array}
            \right)^{-1} \\
            \left(
            \begin{array}{cc}
                2 & 1 \\
                3 & 4
            \end{array}
            \right)
            &=&
            \left(
            \begin{array}{cc}
                1 &  1 \\
                3 & -1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                5 & 0 \\
                0 & 1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                \frac{1}{4} &  \frac{1}{4} \\
                \frac{3}{4} & -\frac{1}{4}
            \end{array}
            \right)
        \end{eqnarray}
        $$
        - 分解することのメリット
            - 行列の累乗の計算が容易になる
            - 行列自体の特徴がわかる
    - 特異値分解（SVD）
        - 固有値分解が$n$次正方行列のみを対象とするのに対し、特異値分解は$(m,n)$型の行列
        $$
        \mathbf{M}=\left(
        \begin{array}{cccc}
            m_{11} & m_{12} & \ldots & m_{1n} \\
            m_{21} & m_{22} & \ldots & m_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            m_{m1} & m_{m2} & \ldots & m_{mn}
        \end{array}
        \right)
        $$
        を対象とする
        - 条件
        $$
        \begin{eqnarray}
            \mathbf{M}\vec{v}&=&\sigma\vec{u} \\
            \mathbf{M}^{\mathrm{T}}\vec{u}&=&\sigma\vec{v}
        \end{eqnarray}
        $$
        を満たす特殊な単位ベクトル$\vec{v},\vec{u}$があるとき、特異値分解が可能
        - 特異値の求め方
        <div style="overflow: auto hidden;">
        $$
        \begin{array}{rcl|rcl|rcccl}
            \mathbf{M}\mathbf{V}&=&\mathbf{U}\mathbf{S} & \mathbf{M}^{\mathrm{T}}\mathbf{U}&=&\mathbf{V}\mathbf{S}^{\mathrm{T}} &
            \mathbf{V}\mathbf{V}^{\mathrm{T}}
            &=&\mathbf{V}^{\mathrm{T}}\mathbf{V}&=&\mathbf{I} \\
            \mathbf{M}&=&\mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}} & \mathbf{M}^{\mathrm{T}}&=&\mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}} &
            \mathbf{U}\mathbf{U}^{\mathrm{T}}
            &=&\mathbf{U}^{\mathrm{T}}\mathbf{U}&=&\mathbf{I} \\
        \end{array}
        $$
        </div>
        $\mathbf{V}$と$\mathbf{U}$は直交行列、$\mathbf{I}$は単位行列であり、これらの積は
        <div style="overflow: auto hidden;">
        $$
        \begin{array}{rcl|rcl}
            \mathbf{M}\mathbf{M}^{\mathrm{T}}
            &=&\mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}}
            \mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}} &
            \mathbf{M}^{\mathrm{T}}\mathbf{M}
            &=&\mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}}
            \mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}} \\
            &=&\mathbf{U}\mathbf{S}\mathbf{I}
            \mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}} &
            &=&\mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{I}
            \mathbf{S}\mathbf{V}^{\mathrm{T}} \\
            &=&\underline{\mathbf{U}\mathbf{S}\mathbf{S}^{\mathrm{T}}}
            \mathbf{U}^{\mathrm{T}} &
            &=&\mathbf{V}
            \underline{\mathbf{S}^{\mathrm{T}}\mathbf{S}\mathbf{V}^{\mathrm{T}}}
        \end{array}
        $$
        </div>となる
        
            - $\mathbf{M}\mathbf{M}^{\mathrm{T}}$は$m$次正方行列、$\mathbf{M}^{\mathrm{T}}\mathbf{M}$は$n$次正方行列となる
            - $\mathbf{M}\mathbf{M}^{\mathrm{T}}$を固有値分解すれば対角行列$\mathbf{S}\mathbf{S}^{\mathrm{T}}$の対角成分から特異値の2乗$\sigma^{2}$、$m$次正方行列$\mathbf{U}$から$m$次元列ベクトルの左特異ベクトル$\vec{u}$が求められる
            - $\mathbf{M}^{\mathrm{T}}\mathbf{M}$を固有値分解すれば対角行列$\mathbf{S}^{\mathrm{T}}\mathbf{S}$の対角成分から特異値の2乗$\sigma^{2}$、$n$次正方行列$\mathbf{V}^{\mathrm{T}}$から$n$次元列ベクトルの右特異ベクトル$\vec{v}$が求められる
            - 左右両特異$\vec{u},\vec{v}$ベクトルは単位ベクトルから作らなければならない
            - 求めた特異値の2乗の平方根の絶対値$|\sqrt{\sigma^{2}}|=\sigma$を用いて、それを対角成分とした対角行列$\mathbf{S}$を求める
            - 最終的に次の式が求まる
            $$
            \mathbf{M}=\mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}}
            $$
    - 或る行列から$\mathbf{\Lambda}$や$\mathbf{S}$のような対角行列、又はそれに準ずる成分構成の行列（例: ジョルダン標準形）を抽出することを「行列の標準化」という
    - 主な用途
        - データ量を減らす
            - [主成分分析](#第4章-:-主成分分析)による次元削減

## 第2章 : 確率・統計
- 集合
    - ものの集まり<br />
    - 或る集合を数学的には
    $$
    S=\{a,b,c,d,e,f,g\}
    $$
    と表現する。このような記法を「外延」という
        - $a$は集合$S$を構成する「要素」（元）の一つ
        $$
        a \in S
        $$
        - $h$は集合$S$を構成する要素ではない
        $$
        h \notin S
        $$
        - 内部に入れ子の集合$M=\{c,d,g\}$がある
        $$
        M \subset S
        $$
    - 確率・統計でいう「事象」は集合として取り扱うことができる
    - 全領域を網羅する集合$U$を「全集合」という
        - ありとあらゆる要素は全て全集合$U$に属すということが暗黙の前提
        $$
        X=\{x \mid x \in U\}
        $$
            - $\mid$記号の後ろに条件を付した記法を「内包」という
        - ありとあらゆる集合は全て全集合$U$の入れ子の集合であることが暗黙の前提
        $$
        X \subset U
        $$
    - 和集合
    $$
    A \cup B = \{x \mid x \in A \,\lor\, x \in B\}
    $$
        - 集合$A$と集合$B$を合わせた領域の集合
        - 要素$x$は集合$A$又は集合$B$に属す  
    - 共通部分
    $$
    A \cap B = \{x \mid x \in A \,\land\, x \in B\}
    $$
        - 集合$A$と集合$B$に共通する領域の集合（積集合ではない）
        - 要素$x$は集合$A$かつ集合$B$に属す
    - 絶対補
    $$
    \begin{eqnarray}
        U \backslash A = \overline{A} &=& \{x \mid x \notin A \,\land\, x \in U\} \\
        &=& \{x \mid x \notin A\}
    \end{eqnarray}
    $$
        - 集合$A$の領域を除く全集合
        - 要素$x$は（全集合$U$に属しつつ）集合$A$に属さない
    - 相対補
    $$
    B \backslash A = \{x \mid x \notin A \,\land\, x \in B\}
    $$
        - 集合$A$の領域を除く集合$B$の領域の集合
        - 要素$x$は集合$B$に属しつつ集合$A$に属さない
    - 可換則
    $$
    A \cap B=B \cap A
    $$
        - 集合$A \cap B$を集合$B \cap A$のように項を交換しても等価
    
    - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
    
    ```python
    # 外延的記法による{1, 5, 8}の集合
    A = {1, 5, 8}
    # 内包的記法による0から9迄の範囲内の奇数要素の集合
    B = {x for x in range(10) if x % 2 != 0}
    
    print("集合A:", A)
    print("集合B:", B)
    print("集合Aと集合Bの和集合:", A | B)
    print("集合Aと集合Bの共通部分:", A & B)
    print("相対補::集合Bに属しつつ集合Aに属さない要素の集合:", B - A)
    ```
    
- 確率
    - 定義<br />或る事象$A$が起こる数に全事象の数を除した連続値
    $$
    P(A)=\frac{n(A)}{n(U)}
    $$
    を事象$A$が起こる「確率」という
        - $P(\cdot)$の値は$0$から$1$迄の連続値
        $$
        0 \leqq P(\cdot) \leqq 1 \hspace{10pt} (P(\cdot) \in \mathbb{R})
        $$
            - 百分率で表記することもある
        - $U$は「全事象」、$n(U)$は「全事象の数」を表わす
            - $n(U)$は$0$より大きい離散値又は連続値
            $$
            n(U) > 0
            $$
        - $A$は$U$に含まれる「事象」の一つ、$n(A)$は「事象$A$が起こる数」を表わす
        $$
        A \subset U
        $$
            - $n(A)$は$0$から$n(U)$迄の範囲内の離散値又は連続値
            $$
            0 \leqq n(A) \leqq n(U)
            $$
    - 全事象$U$の確率$P(U)$は$1$と等価
    $$
    P(U)=\frac{n(U)}{n(U)}=1
    $$
    - $P(A)$の反対の確率は$1-P(A)$と等価
    $$
    \begin{eqnarray}
        P(\overline{A})&=&\frac{n(U)-n(A)}{n(U)}
        &=&\frac{n(U)}{n(U)}&-&\frac{n(A)}{n(U)} \\
        &=&P(U)&-&P(A) \\
        &=&1&-&P(A)
    \end{eqnarray}
    $$
    - 同時確率
        - 定義. 或る事象$A$と、また別の或る事象$B$が共通の事象として同時に起こる確率
        $$
        P(A \cap B)
        $$        
        を「同時確率」という
        - 暗黙の前提である全事象$U$を式に表わすと
        $$
        P(A \cap B)=\frac{P(A \cap B)}{P(U)}
        $$
        となる
            - $P(U)=1$である
            - 一般には$U$が明記されることはない
        - $A$と$B$が互いに因果関係のない独立な事象のとき、互いの事象の確率同士の積
        $$
        P(A \cap B)=P(A)P(B)
        $$
        に変形できる
        - $A \cup B$の中にその共通部分$A \cap B$が存在するとき、$A \cup B$の確率は互いの事象の確率同士の和に対し同時確率を引いたものとなる
        $$
        P(A \cup B)=P(A)+P(B)-P(A \cap B)
        $$
            - 共通部分$A \cap B$が存在しないとき、同時確率は$0$になる
        - $A \cap B$は可換則により
        $$
        A \cap B=B \cap A
        $$
        なので
        $$
        P(A \cap B)=P(B \cap A)
        $$
        を満たす
    - 条件付き確率
        - 定義. 或る条件となる事象$B$が与えられた下で、また別の或る事象$A$起こる確率
        $$
        P(A \mid B)
        $$
        を「条件付き確率」という
        - 暗黙の前提である全事象$U$を考えたとき、事象$A$が起こる確率$P(A)$と、条件となる事象$U$が与えられた下で、事象$A$が起こる確率$P(A \mid U)$は等価である
        $$
        P(A)=P(A \mid U)
        $$
            - 一般には$U$が明記されることはない
        - 同時確率$P(A \cap B)$は、条件付き確率を持つ式に分解できる
        $$
        P(A \cap B)=P(A)P(B \mid A)
        $$
        - $A \cap B$は可換則により
        $$
        A \cap B=B \cap A
        $$
        なので、$P(A)>0$かつ$P(B)>0$のとき
        $$
        \begin{eqnarray}
            P(A \cap B)&=&P(B \cap A)\tag{1} \\
            P(A)P(B \mid A)&=&P(B)P(A \mid B)\tag{2} \\
            P(B)P(A \mid B)&=&P(A \cap B)\tag{3} \\
            P(B \mid A)&=&\frac{P(A \cap B)}{P(A)}\tag{4} \\
            P(A \mid B)&=&\frac{P(A \cap B)}{P(B)}\tag{5}
        \end{eqnarray}
        $$
        を満たす
            - $(4)$の式は、一般に「ベイズ則」（ベイズの定理）の代表式に挙げられる
        - 事象の項が互いに同一のとき、条件付き確率は同時確率以上の数になる
        $$
        \begin{eqnarray}
            P(A \mid B) & \geqq & P(A \cap B) \\
            \frac{P(A \cap B)}{P(B)} & \geqq & \frac{P(A \cap B)}{P(U)}
        \end{eqnarray}
        $$
    - 主客別分類
        - 頻度確率（客観確率）
            - 発生する頻度
            - 例: サイコロ、籤引き
        - ベイズ確率（主観確率）
            - 信念の度合い
            - 直感的には粗筋でレポートを書き進めて行って途中で説明が不足している感じたら遡って補足して行く
- 統計
    - 変数
        - 任意のデータ（標本データ）を表わす添字$i$とデータ数$n$
        $$
        1 \leqq i \leqq n \hspace{10pt} (i, n \in \mathbb{N})
        $$
        - $i$番目のデータの値
        $$
        x_{i} \in \mathbb{R}
        $$
        - データの最小値
        $$
        x_{\mathrm{min}} \in \mathbb{R}
        $$
        - データの最大値
        $$
        x_{\mathrm{max}} \in \mathbb{R}
        $$
        - データの平均
        $$
        \mu = \frac{1}{n}\sum_{i=1}^{n}x_{i}, \hspace{10pt} x_{\mathrm{min}} \leqq \mu \leqq x_{\mathrm{max}} \hspace{10pt} (\mu \in \mathbb{R})
        $$
        - 確率
        $$
        0 \leqq P(\cdot) \leqq 1 \hspace{10pt} (P(\cdot) \in \mathbb{R})
        $$
    - 分類
        - 記述統計（学）
            - 集団の性質を要約（例: 平均）し記述する
            - ビッグデータで扱う
            - 確率変数
                - 事象に紐付けられた数値（例: $100$円コイン）
                - 事象そのものを示すこともある
            - 確率分布
                - 事象の発生する確率の分布
                - 離散値であれば表に示せる
                - 種類
                    - ベルヌーイ分布
                    $$
                    P(x \mid \mu)=\mu^{x}(1-\mu)^{(1-x)}
                    $$
                        - 例: コイントス
                    - マルチヌーイ（カテゴリカル）分布
                        - 例: サイコロ、籤引き
                    - 二項分布
                    $$
                    P(x \mid \lambda, n)=\frac{n!}{x!(n-x)!}\lambda^{x}(1-\lambda)^{n-x}
                    $$
                        - ベルヌーイ分布の多試行版
                        - 山型のグラフが書ける
                    - ガウス分布
                    $$
                    \mathcal{N}(x;\mu,\sigma^{2})=\sqrt{\frac{1}{2\pi\sigma^{2}}\exp\left(-\frac{1}{2\sigma^{2}}(x-\mu)^{2}\right)}
                    $$
                        - 釣鐘型の連続分布
                            - サンプルを多くするほど正規分布に近付く
                            - 座標の全範囲にわたって積分した平面の面積が$1$になる
                            $$
                            \int_{-\infty}^{\infty}\mathcal{N}(x;\mu,\sigma^{2})dx=1
                            $$
            - 期待値
                - その分布に於ける、確率の平均又はありえそうな値を「期待値」という
                - 離散値の期待値
                $$
                \mathrm{E}(f)=\sum_{k=1}^{n}P(X=x_{k})f(X=x_{k})
                $$
                    - $n$は事象$X$の個数
                - 連続値（事象$X$の個数$n$が定まらないとき）の期待値
                $$
                \mathrm{E}(f)=\int P(X=x)f(X=x)dx
                $$
            - 分散と共分散
                - 分散
                $$
                \begin{eqnarray}
                    \mathrm{Var}(f)&=&\mathrm{E}\left(\left(f_{(X-x)}-\mathrm{E}_{(f)}\right)^{2}\right) \\
                    &=&\mathrm{E}\left(f_{(X-x)}^{2}\right)-\left(\mathrm{E}_{(f)}\right)^{2}
                \end{eqnarray}
                $$
                    - データの散らばり具合
                    - データの各々の値が、期待値$\mathrm{E}$からどれだけ乖離しているか平均したもの
                    - 負の値にならない
                        - 絶対値をとるよりも2乗した方が計算コストが小さい
                - 共分散
                $$
                \begin{eqnarray}
                    \mathrm{Cov}(f,g)&=&\mathrm{E}\left(\left(f_{(X-x)}-\mathrm{E}(f)\right)\left(g_{(Y-y)}-\mathrm{E}(g)\right)\right) \\
                    &=&\mathrm{E}(fg)-\mathrm{E}(f)\mathrm{E}(g)
                \end{eqnarray}
                $$
                    - 2つのデータ系列の傾向の違い
                        - 正の値をとれば似た傾向
                        - 負の値をとれば逆の傾向
                        - $0$をとれば関係性に乏しい
            - 分散と標準偏差
                $$
                \sigma=\sqrt{\mathrm{Var}(f)}=\sqrt{\mathrm{E}((f_{(X-x)}-\mathrm{E}_{(f)})^{2})}
                $$
                - 分散$\mathrm{Var}(\cdot)$の平方根を「標準偏差」$\sigma$という
                    - 元のデータと同じ単位
        - 推定統計（学）
            - 調べたい対象である全体の集団を「母集団」という
            - 母集団から抽出したデータを「標本」（サンプル）という
            - 標本を調べ、そこから逆算して母集団が如何なるものか推測する
            - 母集団を特徴づける数を「母数」（パラメータ、平均等）といい、これを統計学的に推測することを「推定」という
                - 母数を1つの値に推定することを「点推定」という
                    - 推定量と推定値
                        - 母数を推定する為に利用する数値の計算方法や計算式のことを「推定量」（推定関数、estimator）という
                        - 実際に試行して行った結果から計算した値を「推定値」（estimate）という
                        - 母集団の（真の）値を$\theta$とすると、推定量又は推定値を$\hat{\theta}$のように表わす
                            - $\hat{\theta}$が関数であれば推定量、具体的な数値であれば推定値である
                    - 標本平均
                        - サンプル数が大きくなれば母集団の値$\theta$に近付くことを「一致性」という
                        - サンプル数がいくらであっても、その期待値$\mathrm{E}(\hat{\theta})$は母集団の値$\theta$と同様であることを「不遍性」という
                        $$
                        \mathrm{E}(\hat{\theta})=\theta
                        $$
                        - 一致性と不偏性両方を満たしていると、母集団を知り易くなる
                    - 標本分散
                    $$
                    \hat{\sigma}^{2}=\frac{1}{n}\sum_{i=1}^{n}(x_{i}-\mu)^{2}
                    $$
                        - 一致性は満たすが、不偏性は満たさない
                    - 不偏分散
                    $$
                    \begin{eqnarray}
                        s^{2}&=&\frac{n}{n-1}\cdot \hat{\sigma}^{2} \\
                        &=&\frac{n}{n-1}\cdot \frac{1}{n}\sum_{i=1}^{n}(x_{i}-\mu)^{2} \\
                        &=&\frac{1}{n-1}\sum_{i=1}^{n}(x_{i}-\mu)^{2}
                    \end{eqnarray}
                    $$
                        - 不偏性をも満たすように標本分散$\hat{\sigma}^{2}$を修正したものを「不偏分散」$s^{2}$という
                - 母数が存在する範囲（区間）を推定することを「区間推定」という

## 第3章 : 情報理論
- 情報科学
    - 情報を如何にして数量化、数理モデル化するか
    - 情報量の増加比率
    $$
    \frac{\Delta w}{w} \hspace{10pt} (\Delta w, w \in \mathbb{R})
    $$
        - $w$は情報量、$\Delta w$は情報の増加量
        - 情報量$10$で情報の増加量$1$
        $$
        \frac{\Delta w}{w}=\frac{1}{10}
        $$
        と、情報量$1$で情報の増加量$1$
        $$
        \frac{\Delta w}{w}=\frac{1}{1}
        $$
        は、両者共に$\Delta w=1$にも拘わらず、後者のほうが視覚的に見分け易い
            - 情報の分かり易さは、情報量の増加比率がカギ
    - 自己情報量
    $$
    I(x)=-\log(P(x))=\log(W(x))
    $$
        - 情報の珍しさを表わす指標
        - $P(\cdot)$は$W(\cdot)$の逆数
        - 対数の底が$2$のとき、単位はビット（bit）
        - 対数の底がネイピアの$e$のとき、単位はナット（nat）
    - シャノンエントロピー
    $$
    \begin{eqnarray}
        H(x)&=&E\left(I(x)\right) \\
        &=&-E\left(\log\left(P(x)\right)\right) \\
        &=&-\sum\left(P(x)\log\left(P(x)\right)\right)
    \end{eqnarray}
    $$
        - 自己情報量$I(\cdot)$の期待値
            - 情報の珍しさの平均
        - 機械学習の際、誤差関数の代替にもなる
    - KLダイバージェンス（カルバック・ライブラー・ダイバージェンス）
    $$
    \begin{eqnarray}
        D_{\mathrm{KL}}(P\|Q)&=&\mathbb{E}_{x \sim P}\left[\log\frac{P(x)}{Q(x)}\right] \\
        &=&\mathbb{E}_{x \sim P}\left[\log P(x)-\log Q(x)\right] \\
        &=&\mathbb{E}_{x \sim P}\left[I\left(Q(x)\right)-I\left(P(x)\right)\right] \\
        &=&\sum_{x}P(x)\left(-\log\left(Q(x)\right)\right)-\left(-\log\left(P(x)\right)\right) \\
        &=&\sum_{x}P(x)\log\frac{P(x)}{Q(x)}
    \end{eqnarray}
    $$
        - 同じ事象・確率変数$x$に於ける異なる確率分布$P,Q$の違いを表わす
        - $Q$は元々考えられていた確率分布、$P$は実際の確率分布
        - 常に正の値
        - ダイバージェンスとは平たくいうと距離のようなもの
            - $D_{\mathrm{KL}}(P\|Q)$と$D_{\mathrm{KL}}(Q\|P)$とでは値が異なるので厳密には距離ではない
        - $\mathbb{E}_{x \sim P}$は期待値
        $$
        E\left(f(x)\right)=\sum_{x}P(x)f(x)
        $$
        を表わす
        - ここから新たに入手できる情報を「情報利得」という
    - 交差エントロピー
    $$
    \begin{eqnarray}
        H(P,Q)&=&H(P)+D_{\mathrm{KL}}(P\|Q) \\
        &=&-\mathbb{E}_{x \sim P}\log Q(x) \\
        &=&-\sum_{x}P(x)\log Q(x)
    \end{eqnarray}
    $$
        - KLダイバージェンス
        $$
        D_{\mathrm{KL}}(P\|Q)=\sum_{x}\underline{P(x)}\left(\underline{-\log\left(Q(x)\right)}\right)-\left(\underline{-\log\left(P(x)\right)}\right)
        $$
        の一部分（下線部）を取り出したもの
            - 距離のようなものですらない
        - $Q$についての自己情報量を$P$の分布で平均している
        - 主にクラス分類用の誤差関数として使用される

# レポート（機械学習）

## 序章 : 機械学習
- 定義
    - コンピュータプログラムは、タスク$T$（アプリケーションにさせたいこと）を性能指標$P$で測定し、その性能が経験$E$（データ）により改善される場合、タスク$T$および性能指標$P$に関して経験$E$から学習（汎化）すると言われている（トム・ミッチェル1997）
    - 人がプログラムするのは経験の仕方ではなく学習の仕方
- パラメータ
    - $w_0$は切片、それ以外の$w_i$は重み付け
    - $\boldsymbol{w}$は要素数$m$のベクトル、$m=1$の場合はスカラー
$$
\begin{cases}
w_1\in\mathbb{R}^1 & (m = 1) \\
\boldsymbol{w} = 
\begin{pmatrix}
w_1 \\
w_2 \\
\vdots \\
w_m
\end{pmatrix}\in\mathbb{R}^m & (m > 1)
\end{cases}
$$
- 説明変数
    - $\boldsymbol{x}$は要素数$m$のベクトル、$m=1$の場合はスカラー
$$
\begin{cases}
x_1\in\mathbb{R}^1 & (m = 1) \\
\boldsymbol{x} = 
\begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_m
\end{pmatrix}\in\mathbb{R}^m & (m > 1)
\end{cases}
$$
- 目的変数
    - 回帰問題
        - $y$はスカラー
$$y\in\mathbb{R}^1$$
    - 分類問題
        - $y$は0か1
$$y\in\{0, 1\}$$
- 平均
$$\bar{x} = \frac{1}{n}\sum_{i=1}^nx_i$$
- データ行列
    - 説明変数（train）の各要素を平均で引く
$$
\bar{\mathbf{X}} =
\begin{pmatrix}
x_1 - \bar{x} \\
x_2 - \bar{x} \\
\vdots \\
x_n - \bar{x}
\end{pmatrix}
$$
- 分散共分散行列
    - データ行列を利用する
$$\Sigma = \mathrm{Var}(\bar{\mathbf{X}}) = \frac{1}{n}\bar{\mathbf{X}}^{\mathrm{T}}\bar{\mathbf{X}}$$
- **シグモイド関数**
    - 入力値$x$はスカラー
    - 出力値$\sigma(\cdot)$は必ず0から1迄の連続値になる∴クラス1に分類される**確率として表現可能**
    - ゲイン$a$の値に応じてシグモイド関数の形が変わる
        - 通常は$a=1$
        - $a=\infty$の場合、形がステップ関数と等しくなる
    - 単調増加関数
$$\sigma(x) = \frac{1}{1 + \exp(-ax)}$$
    - 微分
        - シグモイド関数自身で表現可能
        - 簡単に微分できる形であれば、最大値又は最小値を求める際に有利
$$
\begin{eqnarray}
\frac{\partial \sigma(x)}{\partial x} &=& \frac{\partial}{\partial x}\left(\frac{1}{1 + \exp(-ax)}\right) \\
&=& (-1)\cdot\{1 + \exp(-ax)\}^{-2}\cdot\exp(-ax)\cdot(-a) \\
&=& \frac{a\exp(-ax)}{\{1 + \exp(-ax)\}^2} = \frac{a}{1 + \exp(-ax)}\cdot\frac{1 + \exp(-ax) -1}{1 + \exp(-ax)} \\
&=& a\sigma(x)(1 - \sigma(x))
\end{eqnarray}
$$
- よく使われる確率分布
    - 正規分布
    - t分布
    - ガンマ分布
    - 一様分布
    - ディリクレ分布
    - **ベルヌーイ分布**
        - 確率$p$で1、確率$1-p$で0をとる離散確率分布（例: コイン投げ）
        - ベルヌーイ分布の確率質量関数
$$P(y)=p^y(1 - p)^{1 - y}$$
        - 「生成されるデータ」は分布のパラメータによって異なる
- 同時確率
    - 或るデータが得られた際、それが同時に得られる確率
    - 確率変数は確率的に独立であることを仮定すると、各々の確率の掛け算となる
    - ベルヌーイ分布の確率質量関数の場合
$$P(y_1, y_2, \ldots, y_n; p) = \prod_{i=1}^np^{y_i}(1 - p)^{1 - y_i}$$
    - **$p$を未知のものとすると、尤度関数が求まる**
- **バッチ最急降下法**
    - パラメータ$w_0$の更新量
$$-\eta\frac{\partial L}{\partial w_0} = \eta\sum_{i=1}^n(y_i - x_i)$$
    - パラメータ$w_1$の更新量
$$-\eta\frac{\partial L}{\partial w_1} = \eta\sum_{i=1}^n(y_i - x_i)x_i$$
- **確率的勾配降下法（SGD）**
    - データを一つずつランダム（確率的）に選び学習率$\eta$を掛けたものをパラメータ$\boldsymbol{w}$に足して更新
    - パラメータ$w_0$の更新量
$$-\eta\frac{\partial L}{\partial w_0} = \eta(y_i - x_i)$$
    - パラメータ$w_1$の更新量
$$-\eta\frac{\partial L}{\partial w_1} = \eta(y_i - x_i)x_i$$
    - 特徴
        - 普通の勾配降下法のパラメータ$\boldsymbol{w}$の1回の更新と同じ計算量でパラメータを$n$回更新できる。毎回の更新でデータを一つ又は少量しか見ないので効率的
        - SGDを適用する為には目的関数のパラメータに関する勾配が計算できる必要がある
        - trainデータの順番を変えると結果も変わる可能性がある
        - 学習率$\eta$が小さ過ぎると学習に時間がかかり、局所解に陥り易い
        - 学習率$\eta$が大き過ぎるとパラメータ$\boldsymbol{w}$が発散し易い
- モメンタムありの確率的勾配降下法
    - 今回のSGD更新量$+$前回のSGD更新量に慣性項$\alpha$を掛けたもの
        - 前回のSGD更新量
$$
\left\{
\begin{eqnarray}
\,w_{i\,t} - w_{i\,t-1} \\
i\in\{0, 1\}
\end{eqnarray}
\right.
$$
    - パラメータ$w_0$の更新量
$$
-\eta\frac{\partial L}{\partial w_0} + \alpha(w_{0\,t} - w_{0\,t-1}) 
= \eta(y_i - x_i) + \alpha\eta(y_{i - 1} - x_{i - 1})
$$
    - パラメータ$w_1$の更新量
$$
-\eta\frac{\partial L}{\partial w_1} + \alpha(w_{1\,t} - w_{1\,t-1}) 
= \eta(y_i-x_i)x_i + \alpha\eta(y_{i - 1} - x_{i - 1})x_{i - 1}
$$

- データの分割
    - 学習用データ（train）
        - 機械学習モデルの学習に利用するデータ
        - 学習誤差
    $$MSE_{\mathrm{train}} = \frac{1}{n_{\mathrm{train}}}\sum_{i=1}^{n_{\mathrm{train}}}(\hat{y}^{(\mathrm{train})} - y_i^{(\mathrm{train})})^2$$
    - 検証用データ（test）
        - 学習済みモデルの精度を検証する為のデータ
        - 検証誤差
    $$MSE_{\mathrm{test}} = \frac{1}{n_{\mathrm{test}}}\sum_{i=1}^{n_{\mathrm{test}}}(\hat{y}^{(\mathrm{test})} - y_i^{(\mathrm{test})})^2$$
    - わざわざデータを分割する目的
        - モデルの汎化性能（予測性能）を見る為
    - 汎化性能
        - trainだけでなく、未知の入力データに対する予測性能
        - 検証方法
            - **ホールドアウト法**
                - testを利用して検証
                - データ数が少ないと不利
            - **交差検証（クロスバリデーション）**
                - ホールドアウト法よりも優秀な方法
                - モデルの構築と評価を複数回行う処理
                    - 例えば、データを10分割（$k$分割）した場合、9つ（$k-1$）の集合をtrainに、残りの1つをtestに使用する処理を10回（$k$回）繰り返す。これを10分割交差検証（$k$分割交差検証）と称す
                - CV値が最小のモデルを選択
            - グリッドサーチ
                - 全てのチューニングパラメータ（ハイパーパラメータ）の組み合わせごとにモデルを構築して評価
        - 未学習（underfitting）
            - train誤差が大きいモデル
            - 対策
                - 表現力の高いモデルを利用
        - 過学習（overfitting）
            - train誤差が小さいがtest誤差が大きいモデル
            - trainにきっちりフィットし過ぎていて汎化性能が低い
            - 対策
                - train数を増やす
                - 不要な変数をドロップアウトし表現力を抑止
                - **正則化法**（罰則化法）を利用し表現力を抑止
                    - L1ノルム（**Lasso**）
                        - スパース推定
                            - 幾つかのパラメータを正確に0に推定
                        - 不要な変数等を排除可能
                    - L2ノルム（**Ridge**）
                        - 縮小推定
                            - パラメータを0に近づけるよう推定
- モデルの評価
    - 混同行列（confusion matrix）
        - $tp$（True Positive、真陽性）: 正例と予測して、実際に正例だった数
        - $fp$（False Positive、偽陽性）: 正例と予測したが、実際は負例だった数
            - 確認の工数に時間がかかる
        - $fn$（False Negative、偽陰性）: 負例と予測したが、実際は正例だった数
            - 本当に検知したい脅威が見えなくなる
        - $tn$（True Negative、真陰性）: 負例と予測して、実際に負例だった数

    - 評価方法
        - 正解率（Accuracy）
            - 正解した数$\div$予測対象の全データ数
            - データに偏りがある場合、問題がある
$$
A = \frac{tp + tn}{tp + fn + fp + tn}
$$
        - 再現率（Recall）
            - 正例と予測して、実際に正例だった数$\div$実際正例だった数
            - 適合率とトレードオフ
            - $fp$が多少多くても抜け漏れは少ない際に利用。例えば、実際は癌であるのに癌ではないと誤診することを避けたいタスク
$$
R = \frac{tp}{tp + fn}
$$
        - 適合率（Precision）
            - 正例と予測して、実際に正例だった数$\div$正例と予測した数
            - 再現率とトレードオフ
            - 見逃し（$fn$）が多くても正確な予測をしたい際に利用。例えば、上司からの重要なメールをスパムメールと誤判別されるのを避けたいタスク
$$
P = \frac{tp}{tp + fp}
$$
        - F値
            - 適合率と再現率の調和平均
            - タスクの特性が不明瞭な場合に使用
$$
F = \frac{2}{\frac{1}{P} + \frac{1}{R}} = \frac{2PR}{P + R}
$$

## 第1章 : 線形回帰モデル
- 教師あり学習の回帰手法
- $m$次元の未知の重み付け（パラメータ$\mathbf{w}$）と$m$次元の入力ベクトル（説明変数$\mathbf{x}$）の**線形結合**（内積$+$切片）をスカラーの目的変数$y$として出力するモデル
- **最小二乗法**によりパラメータ$\mathbf{w}$を推定
    - 最小二乗法
        - $y_{\mathrm{train}}$とモデルの出力値$y$の二乗誤差を最小化すること
- 入力
    - 各要素を説明変数又は特徴量と呼ぶ。
    - 要素数を$m$とする。
        - $m=1$の場合はスカラーで表現し、**単回帰**と称す
        - $m>1$の場合はベクトルで表現し、**重回帰**と称す
$$
\begin{cases}
x_1\in\mathbb{R}^1 & (m = 1) \\
\boldsymbol{x} = 
\begin{pmatrix}
x_1 \\
x_2 \\
\vdots \\
x_m
\end{pmatrix}\in\mathbb{R}^m & (m > 1)
\end{cases}
$$
- 出力
    - 目的変数、スカラー
$$y\in\mathbb{R}^1$$
- 特徴
    - 訓練データが多いほど表現力が上がり、過学習が起き難い
    - 外れ値に敏感
    - 説明変数間に相関があると良い推定ができない可能性がある
    - モデルの表現力を下げた方が良い場合もある
    - モデルの表現力が大きくなると、バイアスは小さくなるがバリアンスは大きくなる
	    - バイアスとバリアンスはトレードオフ
- 学習と予測
    - 線形単回帰モデル
        - 説明変数が1次元（$m=1$）
        - データは回帰直線に誤差$\varepsilon$が加わり観測されていると仮定
        - パラメータ$w_0$は切片に相当
        - パラメータ$w_1$は回帰係数に相当
        - モデル数式
$$y = w_0 + w_1x_1 + \varepsilon$$
        - 複数のデータをモデル数式に当てはめると連立方程式が導出される
$$y_1 = w_0 + w_1x_1 + \varepsilon_1$$
$$y_2 = w_0 + w_1x_2 + \varepsilon_2$$
$$\vdots$$
$$y_n = w_0 + w_1x_n + \varepsilon_n$$
    - 線形重回帰モデル
        - 説明変数が多次元（$m>1$）
        - パラメータ$w_0$は切片に相当
        - パラメータ$w_i (i > 0)$は回帰係数に相当
        - データは回帰曲面に誤差$\varepsilon$が加わり観測されていると仮定
        - モデル数式
$$
\begin{eqnarray}
y &=& w_0 + \sum_{i=1}^nw_ix_i+\varepsilon \\
&=& w_0 + w_1x_1 + w_2x_2 + \ldots + w_nx_n + \varepsilon
\end{eqnarray}
$$
        - パラメータ$\boldsymbol{w}$と説明変数$\boldsymbol{x}$の線形結合
$$
\begin{eqnarray}
\hat{y} &=& \boldsymbol{w}^{\mathrm{T}}\boldsymbol{x} + w_0 \\
&=& \sum_{j=1}^mw_jx_j + w_0
\end{eqnarray}
$$
        - **回帰係数$w$の推定は最小二乗法による解と最尤法による解が一致**

## 第2章 : 非線形回帰モデル
- 教師あり学習の回帰手法
- 線形回帰モデルの手法に更に基底関数を混ぜて線形結合する**基底展開法**を使用
    - 未知のパラメータ$w$は線形回帰モデルと同様に最小二乗法や最尤法により推定
$$y_i = w_0 + \sum_{i=1}^mw_j\phi_j(x_i)+\varepsilon_i$$
- よく使われる基底関数
    - 多項式関数
$$\phi_j = x^j$$
    - ガウス型基底関数
$$\phi_j(x) = \exp\left\{\frac{(x - \mu_j)^2}{2h_j}\right\}$$
    - 2次元ガウス型基底関数
$$\phi_j(x) = \exp\left\{\frac{(x - \mu_j)^{\mathrm{T}}(x - \mu_j)}{2h_j}\right\}$$
    - スプライン関数
- 基底展開法も線形回帰と同じ枠組みで推定可能

## 第3章 : ロジスティック回帰モデル
- 教師あり学習の**分類**手法
- パラメータ$\mathbf{w}$と説明変数$\mathbf{x}$の線形結合をシグモイド関数$\sigma(\cdot)$に入力
$$\sigma\left(\sum_{j=1}^mw_jx_j + w_0\right)$$
- 説明変数$\mathbf{x}$が与えられた際に$Y=1$となる確率
$$P(Y=1\mid \mathbf{x})$$
    - シグモイド関数と対応
$$P(Y=1\mid \mathbf{x}) = \sigma\left(\sum_{j=1}^mw_jx_j + w_0\right)$$
- クラス0かクラス1への分類は**ベルヌーイ分布**に準拠
    - 説明変数$\mathbf{x}$が与えられた際に$Y=\{0,1\}$となる確率のベクトル
$$
\begin{eqnarray}
\mathbf{y}_{\mathrm{proba}}
&=&
\begin{pmatrix}
P(Y=0\mid\mathbf{x}) \\
P(Y=1\mid\mathbf{x})
\end{pmatrix}
&=&
\begin{pmatrix}
1 - \sigma\left(\sum_{j=1}^mw_jx_j + w_0\right) \\
\sigma\left(\sum_{j=1}^mw_jx_j + w_0\right)
\end{pmatrix}
\end{eqnarray}
$$
- 学習
    - 確率分布はベルヌーイ分布を利用
    - **最尤推定**によりパラメータ$\mathbf{w}$を推定
        - 最尤推定
            - 尤度関数の最大化
            - $\mathbf{w}$→$p$→確率$P$の順に求まる
$$P(Y=y_1\mid x_1) = p_1^{y_1}(1 - p_1)^{1 - y_1} = \sigma(\mathbf{w}^{\mathrm{T}}x_1)^{y_1}(1-\sigma(\mathbf{w}^{\mathrm{T}}x_1))^{1-y_1}$$
$$P(Y=y_2\mid x_2) = p_2^{y_2}(1 - p_2)^{1 - y_2} = \sigma(\mathbf{w}^{\mathrm{T}}x_2)^{y_2}(1-\sigma(\mathbf{w}^{\mathrm{T}}x_2))^{1-y_2}$$
$$\vdots$$
$$P(Y=y_n\mid x_n) = p_1^{y_n}(1 - p_n)^{1 - y_n} = \sigma(\mathbf{w}^{\mathrm{T}}x_n)^{y_n}(1-\sigma(\mathbf{w}^{\mathrm{T}}x_n))^{1-y_n}$$
    - **尤度関数$L(\cdot)$を最大**とする未知のパラメータ$\mathbf{w}$を探索
        - ベルヌーイ分布の確率質量関数の同時確率を利用
$$
\begin{eqnarray}
P(y_1, y_2, \ldots, y_n \mid w_1, w_2, \ldots, w_n) &=& \prod_{i=1}^n p_i^{y_i} (1 - p_i)^{1-y_i}  \\
&=& \prod_{i=1}^n \sigma(\mathbf{w}^{\mathrm{T}}x_i)^{y_i}(1 - \sigma(\mathbf{w}^{\mathrm{T}}x_i))^{1-y_i} \\
&=& L(\mathbf{w})
\end{eqnarray}
$$
    - **尤度関数が最大となる点と対数尤度関数が最大となる点が一致**する性質を利用
        - 対数をとることで微分の計算が簡単になる
        - 最小二乗法の最小化と合わせる為、尤度関数にマイナスを掛けたものを$E(\cdot)$とし、最大化問題を最小化問題とする
$$
\begin{eqnarray}
E(w_0, w_1, \ldots, w_m) &=& -\log L(w_0, w_1, \ldots, w_m)  \\
&=& \sum_{i=1}^n\{y_i\log p_i+(1-y_i)\log (1-p_i)\}
\end{eqnarray}
$$
- 特徴
    - 表現力の高いモデルに対して正則化すると決定境界は滑らかになる
    - 訓練データが多いほど表現力が上がり、過学習が起き難い

## 第4章 : 主成分分析
- 教師なし学習
- 次元圧縮　高次元の情報の損失をなるべく小さくなるように低次元化する
- 線形変換後のベクトル
    - 主成分ではパラメータ$\mathbf{a}$とする
    - $j$を軸の添え字とする
- 線形変換後の変数の分散が最大となる射影軸を探索
$$
\mathbf{s}_j=
\begin{pmatrix}
s_{1j} \\
s_{2j} \\
\vdots \\
s_{nj}
\end{pmatrix}
=\bar{X}\mathbf{a}_j\in\mathbb{R}^m
$$
- 線形変換後の分散
    - 書き下すと、データ行列$\bar{X}$とその転置行列$\bar{X}^{\mathrm{T}}$の計算式が現れる
$$
\begin{eqnarray}
\mathrm{Var}(\mathbf{s}_j) &=& \frac{1}{n}\mathbf{s}_j^{\mathrm{T}}\mathbf{s}_j \\
&=& \frac{1}{n}(\bar{X}\mathbf{a}_j)^{\mathrm{T}}(\bar{X}\mathbf{a}_j) \\
&=& \frac{1}{n}\mathbf{a}_j^{\mathrm{T}}\bar{X}^{\mathrm{T}}\bar{X} \\
&=& \mathbf{a}_j^{\mathrm{T}}\mathrm{Var}(\bar{X})\mathbf{a}_j
\end{eqnarray}
$$
- 学習
    - ノルムが1となる制約を入れ、最適化問題を解く
    - 目的関数
$$\arg \max (\mathbf{a}_j^{\mathrm{T}})\mathrm{Var}(\bar{X})\mathbf{a}_j\in\mathbb{R}^m$$
    - 制約条件
$$(a_j^{\mathrm{T}})a_j = 1$$
    - 解き方
        - ラグランジュ関数$E(\cdot)$を微分して0になる点を探索
        - 制約条件をラグランジュ乗数とする
$$E(\mathbf{a}_j) = a_j^{\mathrm{T}}\mathrm{Var}(\bar{X})\mathbf{a}_j - \lambda(\mathbf{a}_j^{\mathrm{T}}\mathbf{a}_j - 1)$$
        - $E(\mathbf{a}_j)$を$a$で偏微分
$$\frac{\partial E(\mathbf{a}_j)}{\partial \mathbf{a}_j} = 2\mathrm{Var}(\bar{X})a_j - 2\lambda \mathbf{a}_j = 0$$
- 特徴
    - 解は分散共分散行列の**固有値**と**固有ベクトル**
$$\mathrm{Var}(\bar{X})a_j = \lambda \mathbf{a}_j$$
    - **射影先の分散は固有値と一致**
$$
\begin{eqnarray}
\mathrm{Var}(\mathbf{s}_1) &=& \mathbf{a}_1^{\mathrm{T}}\mathrm{Var}(\bar{X})a_1 \\
&=& \lambda_1\mathbf{a}_1^{\mathrm{T}}\mathbf{a}_1 \\
&=& \lambda_1
\end{eqnarray}
$$
    - 固有値は必ず**0以上**かつ固有ベクトルは**直行**
        - 例えば固有値を2つ求めた場合、それぞれ第1主成分、第2主成分と称す
    - 元データの総分散は主成分の総分散に等しい
$$\mathrm{V}_{\mathrm{total}} = \sum_{i=1}^m\lambda_i$$
- 寄与率
    - 次元圧縮した際の第$k$主成分が持つ情報量の割合
    - 第$k$主成分の分散$\div$主成分の総分散
$$c_k = \frac{\lambda_k}{\sum_{i=1}^m\lambda_i}$$
- 累積寄与率
    - 第1-$k$主成分迄次元圧縮した際の情報損失量の割合
    - 第1-$k$主成分の分散$\div$主成分の総分散
$$r_k = \frac{\sum_{j=1}^k\lambda_j}{\sum_{i=1}^m\lambda_i}$$

## 第7章 : サポートベクターマシーン
- 教師あり学習
- 線形モデルの**正負**で2値分類
    - パラメータ$\mathbf{w}$、説明変数$\mathbf{x}$、切片$b$で線形結合したものを符号関数$\mathrm{sign}(\cdot)$に入力
$$
\mathrm{sign}\left(\sum_{j=1}^mw_jx_j + b\right) =
\begin{cases}
+1 & (y > 0) \\
-1 & (y < 0)
\end{cases}
$$
    - 決定境界はいくつも考えられる
- 学習
    - マージンが最大となる直線（線形判別関数）を探索
$$\max_{\mathbf{w},b}[\min_{i}\frac{t_i\left(\sum_{j=1}^mw_jx_j + b\right)}{\|\mathbf{w}\|}]$$
        - マージン
            - 決定境界から最近傍データ迄のユークリッド距離
$$\min_{i}\frac{t_i\left(\sum_{j=1}^mw_jx_j + b\right)}{\|\mathbf{w}\|}$$
            - マージン上のデータはサポートベクター（サポートベクトル）
                - 予測に影響を与える
$$t_i\left(\sum_{j=1}^mw_jx_j + b - 1\right) = 0$$
            - 目的関数
$$\max_{\mathbf{w},b}\frac{1}{\|\mathbf{w}\|}$$
            - ハイパーパラメータ**C**によるマージンの大きさ調整
                - ソフトマージンSVMでよく使われる
                - Cの値を大きくした場合
                    - マージンが小さくなる
                    - 誤分類に狭量になる
                    - 汎化性能が小さくなる
                    - 過学習が起き易くなる
                - Cの値を小さくした場合
                    - マージンが大きくなる
                    - 誤分類に寛容になる
                    - 汎化性能が大きくなる
                    - 過学習が起き難くなる
    - 線形分離を試行
        - 線形分離不可能な場合は非線形カーネルによって特徴空間に写像して分離（**カーネルトリック**）

- **シグモイド関数**
    - 入力値$x$はスカラー
    - 出力値$\sigma(\cdot)$は必ず0から1迄の連続値になる∴クラス1に分類される**確率として表現可能**
    - ゲイン$a$の値に応じてシグモイド関数の形が変わる
        - 通常は$a=1$
        - $a=\infty$の場合、形がステップ関数と等しくなる
    - 単調増加関数
$$\sigma(x) = \frac{1}{1 + \exp(-ax)}$$

````python
def sigmoid(x: float, gain: float=1.0) -> float:
    """シグモイド関数
    
    Extended description of function.
    
    Parameters
    ----------
    x : float
        入力値を指定します。
    gain : float
        ゲインを指定します。（初期値: 1.0）
        例えば、float('inf') と指定すると関数の振舞がステップ関数になります。
    Returns
    -------
    float
        0.0から1.0迄の連続値を出力します。非数は0.0と看做します。
    """
    sigma = 1 / (1 + np.exp(-gain*x))
    # 非数は0.0と看做す
    return 0.0 if np.isnan(sigma) else sigma

help(sigmoid)
````

- パラメータ$\boldsymbol{w}$と説明変数$\boldsymbol{x}$の線形結合をシグモイド関数$\sigma(\cdot)$に入力
$$\sigma\left(\sum_{j=1}^mw_jx_j + w_0\right)$$
- 説明変数$\boldsymbol{x}$が与えられた際に$Y=1$となる確率
$$P(Y=1\mid \boldsymbol{x})$$
    - シグモイド関数と対応
$$P(Y=1\mid \boldsymbol{x}) = \sigma\left(\sum_{j=1}^mw_jx_j + w_0\right)$$
- クラス0かクラス1への分類は**ベルヌーイ分布**に準拠
    - 説明変数$\boldsymbol{x}$が与えられた際に$Y=\{0,1\}$となる確率のベクトル
$$
\boldsymbol{y}_{proba} = 
\begin{pmatrix}
P(Y=0\mid \boldsymbol{x}) \\
P(Y=1\mid \boldsymbol{x})
\end{pmatrix} = 
\begin{pmatrix}
1 - \sigma\left(\sum_{j=1}^mw_jx_j + w_0\right) \\
\sigma\left(\sum_{j=1}^mw_jx_j + w_0\right)
\end{pmatrix}
$$

````python
# パラメータと説明変数の線形結合をシグモイド関数に入力
proba = sigmoid(model3.coef_ @ vec_x + model3.intercept_)[0]
proba
````

````python
# 乗客の生死の確率を予測
# model3.predict_proba([vec_x])[0] と等価
vec_y_proba = np.array([1 - proba, proba])
vec_y_proba
````

- 予測
    - 予測するデータ点との、距離が最も近い$k$個の、訓練データのラベルの最頻値を割り当てる。

````python
def distance(x1, x2):
    return np.sum((x1 - x2)**2, axis=1)

def knc_predict(n_neighbors, x_train, y_train, X_test):
    y_pred = np.empty(len(X_test), dtype=y_train.dtype)
    for i, x in enumerate(X_test):
        distances = distance(x, X_train)
        nearest_index = distances.argsort()[:n_neighbors]
        mode, _ = stats.mode(y_train[nearest_index])
        y_pred[i] = mode
    return y_pred

def plt_resut(x_train, y_train, y_pred):
    xx0, xx1 = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
    xx = np.array([xx0, xx1]).reshape(2, -1).T
    plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train)
    plt.contourf(xx0, xx1, y_pred.reshape(100, 100).astype(dtype=np.float), alpha=0.2, levels=np.linspace(0, 1, 3))
````

# レポート（深層学習・day1）

## Introduction : ディープラーニング（深層学習）
- 定義
    - ディープラーニングとは機械学習の一種であり、入力層、多層の中間層、及び出力層で構成された人工ニューラルネットワーク（ANN）上に於いて、出力値と正解ラベルの誤差$E(\mathbf{y}_{nk})$を最小化する最適なパラメータ$\mathbf{w}$（重み$\mathbf{W}$とバイアス$\mathbf{b}$）を探索すること
        - 誤差関数$E(\cdot)$は、損失関数ともいう
        - 認識精度を指標にしないのは、微分がほとんどの場所で0になり、パラメータ$\mathbf{w}$更新が不可能になるから
    - 他の機械学習と大きく異なる点の一つはend-to-endであること
- 全体像
    - 各要素のインデックス（添字）
        - 入力層ノード
$$i\in\left\{\texttt{1}, \cdots, I\right\}\subset\mathbb{N}$$
        - 中間層ノード
$$j\in\left\{\texttt{1}, \cdots, J\right\}\subset\mathbb{N}$$
        - 出力層ノード
$$k\in\left\{\texttt{1}, \cdots, K\right\}\subset\mathbb{N}$$
        - 入力層以外の層
$$l\in\left\{\texttt{1}, \cdots, L\right\}\subset\mathbb{N}$$
        - 訓練データ
$$n\in\left\{\texttt{1}, \cdots, N\right\}\subset\mathbb{N}$$
        - 試行回数
$$t\in\left\{\texttt{1}, \cdots, T\right\}\subset\mathbb{N}$$
    - **事前に用意する情報**
$$
\begin{array}{rcl}
\text{input}:
\mathbf{x}_n
&=&
\left[
    \begin{array}{ccc}
        x_{n\texttt{1}} & \cdot\cdot & x_{ni}
    \end{array}
\right]
\\
\text{training data}:
\mathbf{d}_n
&=&
\left[
    \begin{array}{ccc}
        d_{n\texttt{1}} & \cdot\cdot & d_{ni}
    \end{array}
\right]
\end{array}
$$
    - **多層ネットワークのパラメータ（学習対象）**
$$
\mathbf{w}^{(l)}
\left\{
\begin{array}{rcl}
\text{Weight}:
\mathbf{W}^{(l)}
&=&
\left[
    \begin{array}{ccc}
        w_{\texttt{11}}^{(l)} & \cdots & w_{\texttt{1}l}^{(l)} \\
        \vdots & \ddots & \vdots \\
        w_{J\texttt{1}}^{(l)} & \cdots & w_{Jl}^{(l)}
    \end{array}
\right]
\\
\text{bias}:
\mathbf{b}^{(l)}
&=&
\left[
\begin{array}{ccc}
    b_{\texttt{1}}^{(l)} & \cdots & b_{l}^{(l)}
\end{array}
\right]
\end{array}
\right.
$$
    - 活性化関数
$$
\mathbf{f}^{(l)}(u^{(l)})
=\left[
\begin{array}{ccc}
    f^{(l)}(u_{\texttt{1}}^{(l)}) & \cdot\cdot & f^{(l)}(u_{J}^{(l)})
\end{array}
\right]
$$
    - 中間層出力
$$
\mathbf{z}^{(l)}
=\left[
\begin{array}{ccc}
    z_{\texttt{1}}^{(t)} & \cdot\cdot & z_{K}^{(t)}
\end{array}
\right]
=\mathbf{f}^{(l)}(u^{(l)})
$$
    - 総入力
$$
\mathbf{u}^{(l)}
=\begin{cases}
    \mathbf{W}^{(l)}\mathbf{x} + \mathbf{b}^{(l)} & \quad (l=\texttt{1}) \\
    \mathbf{W}^{(l)}\mathbf{z}^{(l - \texttt{1})} + \mathbf{b}^{(l)} & \quad (\text{otherwise})
\end{cases}
$$
    - 出力
$$
\mathbf{y}_{n}^{(t)}
=\left[
\begin{array}{ccc}
    y_{n\texttt{1}}^{(t)} & \cdot\cdot & y_{nK}^{(t)}
\end{array}
\right]
=\mathbf{z}^{(L)}
$$
    - 誤差関数
$$
E_{n}(\mathbf{w})
$$
    - 確率的勾配降下法（SGD）
$$
\left\{
\begin{eqnarray}
    \mathbf{w}^{(t + \texttt{1})} &=& \mathbf{w}^{(t)} - \varepsilon\nabla E_{n}(\mathbf{w}) \\
    \nabla E_{n}(\mathbf{w}) &=& \frac{\partial E}{\partial \mathbf{w}}
\end{eqnarray}
\right.
$$

## Section2 : 活性化関数
- 総入力$\mathbf{u}$を入力値とする非線形関数
$$
\mathbf{f}^{(l)}(u^{(l)})=\left[
\begin{array}{ccc}
    f^{(l)}(u_{\texttt{1}}^{(l)}) & \cdot\cdot & f^{(l)}(u_{J}^{(l)})
\end{array}
\right]
$$
- 中間層$(l \leq L-1)$用の活性化関数
    - ReLU関数（正規化線形関数）
$$
f(x)=\begin{cases}
    x & (x > 0) \\
    0 & (x \leq 0)
\end{cases}
$$
        - 現在最も使用されている
        - 勾配消失問題の回避とスパース化に貢献
    - シグモイド関数（ロジスティック関数）
$$f(x) = \frac{1}{1 + \exp(-x)}$$
        - 勾配消失問題が起こり易い
    - ステップ関数
$$f(x) = \frac{1}{1 + \exp(-\infty x)}$$
$$
f(x)=\begin{cases}
    1 & (x \geq 0) \\
    0 & (x < 0)
\end{cases}
$$
        - ON/OFFしか出力しない
        - 線形分離可能な問題しか解けない
        - 現在ディープラーニングでは基本的に用いられない
    - <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>
    
    ```python
    import numpy as np

    # 中間層の活性化関数の定義
    def sigmoid(x):
           """シグモイド関数 (ロジスティック関数)"""
           return 1/(1 + np.exp(-x))

    def relu(x):
           """ReLU関数 (正規化線形関数)"""
           return np.maximum(0, x)

    def step_function(x):
           """ステップ関数 (閾値0の階段関数)"""
           return np.where( x > 0, 1, 0)

    # ハイパーパラメータ
    ## 中間層の活性化関数
    hidden_f = relu
    
    # 入力
    x = np.array([1.0, 0.5])
    # パラメータの初期設定
    W1 = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    b1 = np.array([0.1, 0.2, 0.3])
    # 総入力
    u1 = np.dot(x, W1) + b1
    # 中間層出力
    z1 = hidden_f(u1)

    print("中間層の活性化関数:", hidden_f.__doc__.splitlines()[0])
    ```

## Section3 : 出力層
- 値の強弱と確率出力の観点から出力層と中間層で利用される活性化関数が異なる
- 出力層の活性化関数は多クラス分類で主にソフトマックス関数が使われる

||回帰|二値分類|多クラス分類|
|:-:|:-:|:-:|:-:|
|**活性化関数**|<p style="width: 16em; text-align: center;">恒等写像</p><p style="height: 2em; width: 16em; text-align: center;">$f(u) = u$</p>|<p style="width: 16em; text-align: center;">シグモイド関数</p><p style="height: 2em; width: 16em; text-align: center;">$f(u) = \frac{1}{1 + \exp(-u)}$</p>|<p style="width: 16em; text-align: center;">ソフトマックス関数</p><p style="height: 2em; width: 16em; text-align: center;">$f(i, u) = \frac{\exp(u_{i})}{\sum_{k=1}^{K}\exp(u_{k})}$</p>|
|**誤差関数**|二乗誤差|交差エントロピー|交差エントロピー|

- 訓練データサンプルあたりの誤差
    - 二乗誤差
$$
E_{n}(\mathbf{w}) = \frac{1}{2}\sum_{i=1}^{I}(\mathbf{y}_{n} - \mathbf{d}_{n})^{2}
$$
    - 交差エントロピー
$$
E_{n}(\mathbf{w}) = -\sum_{i=1}^{I}d_{i}\log y_{i}
$$
- 学習サイクルあたりの誤差
$$
E_{n}(\mathbf{w}) = \sum_{n=1}^{N}E_{n}
$$

## Section4 : 勾配降下法
- 誤差をパタメータで微分
$$
\nabla E
=\frac{\partial E}{\partial \mathbf{w}}
=\left[
\begin{array}{ccc}
    \frac{\partial E}{\partial w_{1}} & \cdot\cdot & \frac{\partial E}{\partial w_{M}}
\end{array}
\right]
$$
- 勾配降下法（最急降下法）
    - 全サンプルの平均誤差
$$
\mathbf{w}^{(t + \texttt{1})} = \mathbf{w}^{(t)} - \varepsilon\nabla E
$$
- 確率的勾配降下法（SGD）
    - ランダムに抽出したサンプルの誤差
    - 計算コスト軽減、局所解収束リスク軽減、オンライン学習可能
$$
\mathbf{w}^{(t + \texttt{1})} = \mathbf{w}^{(t)} - \varepsilon\nabla E_{n}
$$
- ミニバッチ勾配降下法
    - ランダムに分割したデータの集合（ミニバッチ$D_{t}$）に属するサンプルの平均誤差
    - SGDの利点を損なわず計算資源を有効活用できる
$$
\left\{
    \begin{eqnarray}
        \mathbf{w}^{(t + \texttt{1})} &=& \mathbf{w}^{(t)} - \varepsilon\nabla E_{t} \\
        E_{t} &=& \frac{1}{N_{t}}\sum_{n\in D_{t}}E_{n} \\
        N_{t} &=& |D_{t}|
    \end{eqnarray}
\right.
$$
- 数値微分
    - 微小な数値$h$を距離として擬似的（非解析的）に微分すること
    - 計算回数が多くなるのがデメリット
$$
\frac{\partial E}{\partial w_{m}}
\approx
\frac{E(w_{m} + h) - E(w_{m} - h)}{2h}
$$

## Section5 : 誤差逆伝播法
- 計算結果を逆方向の層に伝播して解析的に微分計算する手法
- 数値微分のデメリットを回避
- 連鎖率の原理
$$
\frac{\mathrm{d}z}{\mathrm{d}x}=\frac{\mathrm{d}z}{\mathrm{d}t} \cdot \frac{\mathrm{d}t}{\mathrm{d}x}
$$
    - 例. $z=t^2\ ,\ t=x+y$の式の$\frac{\mathrm{d}z}{\mathrm{d}x}$を解くと
$$
\frac{\mathrm{d}z}{\mathrm{d}t}=2t\ ,\ 
\frac{\mathrm{d}t}{\mathrm{d}x}=1
$$
$$
\begin{eqnarray}
\frac{\mathrm{d}z}{\mathrm{d}x}&=&\frac{\mathrm{d}z}{\mathrm{d}t} \cdot \frac{\mathrm{d}t}{\mathrm{d}x} \\
&=&2t \cdot 1 \\
&=&2t \\
&=&2(x+y) \\
\end{eqnarray}
$$
    - 例. 入力層1層、中間層1層、出力層1層のネットワーク構成$(L=2)$の誤差逆伝播法
        - 出力層$(l=2)$のパラメータ$\mathbf{w}^{(2)}$
$$
\frac{\partial E}{\partial \mathbf{w}^{(2)}}
=\frac{\partial E}{\partial \mathbf{y}} \cdot
\frac{\partial \mathbf{y}}{\partial \mathbf{u}^{(2)}} \cdot
\frac{\partial \mathbf{u}^{(2)}}{\partial \mathbf{w}^{(2)}}
$$
            - 例. 誤差関数を二乗誤差
            $$
            E(\mathbf{y}) = \frac{1}{2}\sum_{i=1}^{I}(\mathbf{y}_{n} - \mathbf{d}_{n})^{2}
            =\frac{1}{2}\|\mathbf{y}-\mathbf{d}\|^{2}
            $$
            としたとき、これを$\mathbf{y}$で偏微分すると
            $$
            \frac{\partial E(\mathbf{y})}{\partial \mathbf{y}}=\frac{\partial}{\partial \mathbf{y}}\frac{1}{2}\|\mathbf{y}-\mathbf{d}\|^{2}= \mathbf{y}-\mathbf{d}
            $$
            となるので
            $$
            \frac{\partial E}{\partial \mathbf{w}^{(2)}}
            =\frac{\partial E}{\partial \mathbf{y}} \cdot
            \frac{\partial \mathbf{y}}{\partial \mathbf{u}^{(2)}} \cdot
            \frac{\partial \mathbf{u}^{(2)}}{\partial \mathbf{w}_{ji}^{(2)}}
            =(\mathbf{y}-\mathbf{d}) \cdot
            \left[
            \begin{array}{c}
                0 \\
                \vdots \\
                z_{i} \\
                \vdots \\
                0
            \end{array}
            \right]
            =(y_{j}-d_{j})z_{i}
            $$
            を満たす
            - バイアス$\mathbf{b}^{(2)}$
$$
\frac{\partial E}{\partial \mathbf{b}^{(2)}}
=\frac{\partial E}{\partial \mathbf{y}} \cdot
\frac{\partial \mathbf{y}}{\partial \mathbf{u}^{(2)}} \cdot
\frac{\partial \mathbf{u}^{(2)}}{\partial \mathbf{b}^{(2)}}
$$
        - 中間層$(l=1)$のパラメータ$\mathbf{w}^{(1)}$
$$
\frac{\partial E}{\partial \mathbf{w}^{(1)}}
=\frac{\partial E}{\partial \mathbf{y}} \cdot
\frac{\partial \mathbf{y}}{\partial \mathbf{u}^{(2)}} \cdot
\frac{\partial \mathbf{u}^{(2)}}{\partial \mathbf{z}^{(1)}} \cdot
\frac{\partial \mathbf{z}^{(1)}}{\partial \mathbf{u}^{(1)}} \cdot
\frac{\partial \mathbf{u}^{(1)}}{\partial \mathbf{w}^{(1)}}
$$
            - バイアス$\mathbf{b}^{(1)}$
$$
\frac{\partial E}{\partial \mathbf{b}^{(1)}}
=\frac{\partial E}{\partial \mathbf{y}} \cdot
\frac{\partial \mathbf{y}}{\partial \mathbf{u}^{(2)}} \cdot
\frac{\partial \mathbf{u}^{(2)}}{\partial \mathbf{z}^{(1)}} \cdot
\frac{\partial \mathbf{z}^{(1)}}{\partial \mathbf{u}^{(1)}} \cdot
\frac{\partial \mathbf{u}^{(1)}}{\partial \mathbf{b}^{(2)}}
$$
        - 連鎖率の原理に裏付けられた誤差逆伝播法を用いることによって、解析的な（導関数を用いた）微分が可能
            - 欠点の多い数値微分に頼らずに済む
        -  <div style="vertical-align: middle; display: inline-block;"><a href="https://www.python.org/" target="_blank"><img src="https://www.python.org/static/community_logos/python-logo-generic.svg" alt="Python" title="The Python Programming Language" align="left" width="88" height="26" style="max-height: 26px; height: 26px; display: inline-block; vertical-align: middle;" /></a>コード例</div>

        ```python
        import numpy as np

        # 中間層の活性化関数の定義
        def sigmoid(x):
               """シグモイド関数 (ロジスティック関数)"""
               return 1/(1 + np.exp(-x))

        def relu(x):
               """ReLU関数 (正規化線形関数)"""
               return np.maximum(0, x)

        def step_function(x):
               """ステップ関数 (閾値0の階段関数)"""
               return np.where( x > 0, 1, 0)

        # 出力層の活性化関数の定義
        def identity_function(x):
               """恒等写像"""
               return x

        def softmax(x):
               """ソフトマックス関数"""
               if x.ndim == 2:
                   x = x.T
                   x = x - np.max(x, axis=0)
                   y = np.exp(x) / np.sum(np.exp(x), axis=0)
                   return y.T

               x = x - np.max(x) # オーバーフロー対策
               return np.exp(x) / np.sum(np.exp(x))

        def softmax_with_loss(d, x):
               """ソフトマックスと交差エントロピーの複合関数"""
               y = softmax(x)
               return cross_entropy_error(d, y)

        # 誤差関数の定義
        def mean_squared_error(d, y):
               """平均二乗誤差"""
               return np.mean(np.square(d - y)) / 2

        def cross_entropy_error(d, y):
               """交差エントロピー"""
               if y.ndim == 1:
                   d = d.reshape(1, d.size)
                   y = y.reshape(1, y.size)

               # 訓練データがone-hot-vectorの場合、正解ラベルのインデックスに変換
               if d.size == y.size:
                   d = d.argmax(axis=1)

               batch_size = y.shape[0]
               return -np.sum(np.log(y[np.arange(batch_size), d] + 1e-7)) / batch_size

        # 活性化関数の導関数
        def d_sigmoid(x):
               """シグモイド関数の導関数"""
               dx = (1.0 - sigmoid(x)) * sigmoid(x)
               return dx

        def d_relu(x):
               """ReLU関数の導関数"""
               return np.where( x > 0, 1, 0)

        def d_step_function(x):
               """ステップ関数の導関数"""
               return 0

        def d_mean_squared_error(d, y):
               """平均二乗誤差の導関数"""
               if type(d) == np.ndarray:
                   batch_size = d.shape[0]
                   dx = (y - d)/batch_size
               else:
                   dx = y - d
               return dx

        def d_softmax_with_loss(d, y):
               """ソフトマックスと交差エントロピーの複合導関数"""
               batch_size = d.shape[0]
               if d.size == y.size: # 訓練データがone-hot-vectorの場合
                   dx = (y - d) / batch_size
               else:
                   dx = y.copy()
                   dx[np.arange(batch_size), d] -= 1
                   dx = dx / batch_size
               return dx

        def d_sigmoid_with_loss(d, y):
               """シグモイドと交差エントロピーの複合導関数"""
               return y - d

        # ハイパーパラメータ
        ## 中間層の活性化関数
        hidden_f = relu
        ## 出力層の活性化関数
        last_f = softmax
        ## 誤差関数
        loss_f = cross_entropy_error
        ## 出力層の活性化関数の導関数
        last_d_f = d_softmax_with_loss
        ## 中間層の活性化関数の導関数
        hidden_d_f = d_relu
        ## 学習率
        learning_rate = 0.1

        # サブルーチンの定義
        def init_network():
               """パラメータの初期設定"""
               network = {} 
               network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
               network['b1'] = np.array([0.1, 0.2, 0.3])
               network['W2'] = np.array([[0.1, 0.4, 0.7], [0.2, 0.5, 0.8], [0.3, 0.6, 0.9]])
               network['b2'] = np.array([0.1, 0.2, 0.3])

               return network

        def forward(network, x):
               """順伝播"""
               W1, W2 = network['W1'], network['W2']
               b1, b2 = network['b1'], network['b2']

               # 総入力
               u1 = np.dot(x, W1) + b1
               # 中間層出力
               z1 = hidden_f(u1)
               # 総入力
               u2 = np.dot(z1, W2) + b2
               # 出力
               z2 = last_f(u2)
               y  = z2

               return z1, y

        def backward(x, d, z1, y):
               """誤差逆伝播"""
               print("\n##### 誤差逆伝播開始 #####")
               grad = {}
               W1, W2 = network['W1'], network['W2']
               b1, b2 = network['b1'], network['b2']

               # 出力層でのデルタ
               delta2 = last_d_f(d, y)
               # b2の勾配
               grad['b2'] = np.sum(delta2, axis=0)
               # W2の勾配
               grad['W2'] = np.dot(z1.T, delta2)
               # 中間層でのデルタ
               delta1 = np.dot(delta2, W2.T) * hidden_d_f(z1)
               # b1の勾配
               grad['b1'] = np.sum(delta1, axis=0)
               # W1の勾配
               grad['W1'] = np.dot(x.T, delta1)

               return grad

        # 入力
        x = np.array([1.0, 0.5])
        # パラメータの初期設定
        network = init_network()
        # 順伝播
        z1, y = forward(network, x)

        print("中間層の活性化関数:", hidden_f.__doc__.splitlines()[0])
        print("出力層の活性化関数:", last_f.__doc__.splitlines()[0])
        print("誤差関数:", loss_f.__doc__.splitlines()[0])
        print("出力y:", y)
        for k in range(len(y)):
               print("y_{k:d}を正解とした場合の誤差:".format(k=k+1), loss_f(np.array([k], dtype=int), y))

        # 学習
        ## 正解をy_1(y[0])とする
        k = 0
        d = np.array([k], dtype=int)
        ## 勾配降下法
        ### 誤差逆伝播法によって勾配を求める
        grad = backward(x[np.newaxis, :], d, z1[np.newaxis, :], y[np.newaxis, :])
        ### パラメータに勾配適用
        for key in ('W1', 'W2', 'b1', 'b2'):
               network[key]  -= learning_rate * grad[key]

        ## 勾配適用済みパラメータで再順伝播
        y = forward(network, x)[-1]

        print("\n##### y_{k:d}が正解という前提の元、学習率{lr:g}で学習しました。 #####\n".format(k=k+1, lr=learning_rate))
        print("出力層の活性化関数の導関数:", last_d_f.__doc__.splitlines()[0])
        print("中間層の活性化関数の導関数:", hidden_d_f.__doc__.splitlines()[0])
        print("出力y:", y)
        for k in range(len(y)):
               print("y_{k:d}を正解とした場合の誤差:".format(k=k+1), loss_f(np.array([k], dtype=int), y))
        ```
        
    - 計算前と比較すると、出力値と正解ラベルの誤差$E(\mathbf{y}_{nk})$が小さくなる→ディープラーニングの最終目的でもあるパラメータ$\mathbf{w}$（重み$\mathbf{W}$、バイアス$\mathbf{b}$）の値の最適化が行われた
        - この操作を繰り返して誤差$E(\mathbf{y}_{nk})$を更に小さくして行く

````python
import numpy as np

def chk_network(network):
    """ネットワークを検査
    
    この関数は、指定したネットワーク構成を検査します。
    異常なければ、パラメータと想定ネットワーク構成の情報を
    可視化図と合わせて出力します。
    異常を検知したら、例外（Exception）が発生します。
    """
    print("##### ネットワークの検査 #####")
    L = len(network) / 2
    #print(L)
    L_ = 0
    for elem in network:
        if elem.startswith('W') or elem.startswith('b'):
            L_ += 1
            
    if L != (L_ / 2):
        raise Exception("""ネットワーク構成の異常を検知しました！
        Wとbの件数は互いに等しくなければなりません。""")
        
    L = int(L)
    
    mat_w = np.array([('b' + str(a), 'W' + str(a)) for a in range(1, L+1)])
    illegal_keys = set(network.keys()) - set(mat_w.flatten())
    if len(illegal_keys) != 0:
         raise Exception("""ネットワーク構成の異常を検知しました！
         {}は不正なパラメータ名です。""".format(str(illegal_keys)))
        
    il_conf = dict(n_nodes=0, n_layers=1)
    hl_conf = dict(vec_nodes=[], n_layers=L-1)
    ol_conf = dict(n_nodes=0, n_layers=1)
    
    n_nodes = len(network[mat_w[0][1]])
    if not n_nodes:
        raise Exception("""ネットワーク構成の異常を検知しました！
        パラメータW1の値が不正です。""")
        
    il_conf['n_nodes'] = n_nodes
        
    for b, W in mat_w:
        if n_nodes != len(network[W]):
            raise Exception("""ネットワーク構成の異常を検知しました！
            パラメータ{}の値が不正です。""".format(str(W)))
        
        n_nodes = len(network[b])
        if n_nodes != len(network[W][0]):
            raise Exception("""ネットワーク構成の異常を検知しました！
            パラメータ{}と{}[0]の要素数が不一致です。""".format(str(b), str(W)))
            
        hl_conf['vec_nodes'].append(n_nodes)
            
    # 異常なし
    
    ol_conf['n_nodes'] = hl_conf['vec_nodes'].pop()
    
    print("*** パラメータの値と形状 ***")
    for b, W in mat_w:
        print(b, "=", network[b], network[b].shape)
        print(W, "=")
        print(network[W], network[W].shape)
        print()
        
    print("*** 想定ネットワーク構成 ***")
    print("入力層:", "{n_layers:d}層 {n_nodes:d}ノード".format(**il_conf))
    print("中間層:", "{n_layers:d}層 {vec_nodes}ノード".format(**hl_conf))
    print("出力層:", "{n_layers:d}層 {n_nodes:d}ノード".format(**ol_conf))
    
    # 作図
    #from IPython.display import SVG, display
    
    vec_node_cx = np.linspace(50, 200 * L + 50, L + 1, dtype=int)
    vec_node_stroke = np.full_like(vec_node_cx, 'green', dtype=object)
    vec_node_stroke[0] = 'red'; vec_node_stroke[-1] = 'blue'
    vec_n_nodes = np.array([il_conf['n_nodes']] + hl_conf['vec_nodes'] + [ol_conf['n_nodes']])
    n_nodes_max = vec_n_nodes.max()
    half_max_cy = 50 * (n_nodes_max - 1)
    vec_node_cy_margin = np.array([half_max_cy - (half_max_cy * l / n_nodes_max) for l in vec_n_nodes], \
                                  dtype=int)
    string_builder = []
    string_builder.append('<svg width="{width:d}" height="{height:d}" viewBox="0 0 850 100">'.format( \
                                                                width=200*L+50, height=155*(n_nodes_max)))
    string_builder.append('<marker id="edge" markerUnits="userSpaceOnUse" markerWidth="30" markerHeight="30" ')
    string_builder.append('viewBox="0 0 10 10" refX="9" refY="5" orient="auto">')
    string_builder.append('<polygon points="0,0 5,5 0,10 10,5 " fill="auto"/>')
    string_builder.append('</marker>')
    string_builder.append('<circle cx="6" cy="-220" stroke="red" r="5" fill="none" stroke-width="1" />')
    string_builder.append('<circle cx="6" cy="-180" stroke="green" r="5" fill="none" stroke-width="1" />')
    string_builder.append('<circle cx="6" cy="-140" stroke="blue" r="5" fill="none" stroke-width="1" />')
    string_builder.append('<text x="18" y="-216" fill="red">入力層のノード</text>')
    string_builder.append('<text x="18" y="-176" fill="green">中間層のノード</text>')
    string_builder.append('<text x="18" y="-136" fill="blue">出力層のノード (L={L:d})</text>'.format(L=L))
    sl_node_z_points = []
    for l, node_cx, node_stroke, n_nodes, node_cy_margin in \
    zip(range(L + 1), vec_node_cx, vec_node_stroke, vec_n_nodes, vec_node_cy_margin):
    
        cl_node_z_points = []
        for ppj, node_cy in zip(range(1, n_nodes + 1), (100 * j + node_cy_margin for j in range(n_nodes))):
            cl_node_z_points.append((node_cx + 40, node_cy))
            string_builder.append('<circle cx="{cx:d}" cy="{cy:d}" stroke="{stroke:s}" '.format( \
                                                                                    cx=node_cx, \
                                                                                    cy=node_cy, \
                                                                                    stroke=node_stroke))
            string_builder.append('r="40" fill="none" stroke-width="1" />')
            
            if l != 0:
                string_builder.append( \
                    '<text x="{x:d}" y="{y:d}" fill="black">u<tspan baseline-shift="sub">{ppj:d}'.format( \
                                                                        x=node_cx-36, y=node_cy, ppj=ppj))
                string_builder.append( \
                    '</tspan><tspan fill="{fill:s}" baseline-shift="super">({l:d})</tspan></text>'.format( \
                                                                                fill=node_stroke, l=l))
                string_builder.append( \
                    '<text x="{x:d}" y="{y:d}" fill="black">z<tspan baseline-shift="sub">{ppj:d}'.format( \
                                                                        x=node_cx+2, y=node_cy, ppj=ppj))
                string_builder.append( \
                    '</tspan><tspan fill="{fill:s}" baseline-shift="super">({l:d})</tspan></text>'.format( \
                                                                                fill=node_stroke, l=l))
                string_builder.append( \
                    '<line x1="{x1:d}" y1="{y1:d}" x2="{x2:d}" y2="{y2:d}" stroke="{stroke:s}" />'.format( \
                                    x1=node_cx, y1=node_cy-40, x2=node_cx, y2=node_cy+40, stroke=node_stroke))
                
                x2 = node_cx - 40
                for x1, y1 in sl_node_z_points:
                    #print(x1, y1)
                    string_builder.append('<line x1="{x1:d}" y1="{y1:d}" x2="{x2:d}" y2="{y2:d}" '.format( \
                                                            x1=x1, y1=y1, x2=x2, y2=node_cy))
                    string_builder.append('stroke="black" stroke-width="2" marker-end="url(#edge)" />')
                    
            else:
                string_builder.append( \
'<text x="{x:d}" y="{y:d}" fill="black">x<tspan baseline-shift="sub">{ppj:d}</tspan></text>'.format( \
                                                            x=node_cx-10, y=node_cy, ppj=ppj))
        
        if l != 0:
            string_builder.append('<text x="{x:d}" y="{y:d}" fill="{fill:s}">l={l:d}</text>'.format( \
                                                x=node_cx-10, y=half_max_cy*2+60, fill=node_stroke, l=l))
            string_builder.append('<text x="{x:d}" y="{y:d}" fill="black">f<tspan fill="'.format( \
                                                x=node_cx-24, y=-50))
            string_builder.append( \
'{fill:s}" baseline-shift="super">({l:d})</tspan>(u<tspan fill="{fill:s}" baseline-shift="super">({l:d}'.format( \
                                                fill=node_stroke, l=l))
            string_builder.append(')</tspan>)</text>')
            
            string_builder.append('<text x="{x:d}" y="{y:d}" transform="rotate(90) scale(1,1.8)" '.format( \
                                                x=half_max_cy*2+68, y=int(-sl_node_z_points[0][0]/1.8-6)))
            string_builder.append('font-family="serif" font-size="100" fill="black">}</text>')
            
            string_builder.append( \
                 '<text x="{x:d}" y="{y:d}" fill="black">W<tspan baseline-shift="super" '.format( \
                                                                        x=node_cx-120, y=half_max_cy*2+120))
            string_builder.append( \
'fill="{fill:s}" >({l:d})</tspan>, b<tspan fill="{fill:s}" baseline-shift="super">({l:d}'.format( \
                                                                        fill=node_stroke, l=l))
            string_builder.append(')</tspan></text>')
            
        sl_node_z_points = cl_node_z_points
        
        #print()
    ppj = 1
    for x1, y1 in cl_node_z_points:
        string_builder.append('<line x1="{x1:d}" y1="{y1:d}" x2="{x2:d}" y2="{y2:d}" '.format( \
                                                                        x1=x1, y1=y1, x2=x1+120, y2=y1))
        string_builder.append('stroke="black" stroke-width="2" marker-end="url(#edge)" />')
                    
        string_builder.append( \
'<text x="{x:d}" y="{y:d}" fill="black">y<tspan baseline-shift="sub">{ppj:d}</tspan></text>'.format( \
                                                                        x=x1+130, y=y1, ppj=ppj))
        ppj += 1
        
    string_builder.append('</svg>')
    svg_data = ''.join(string_builder)
    display(SVG(data=svg_data))

def init_network():
    """パラメータの初期設定"""
    network = {} 
    network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
    network['b1'] = np.array([0.1, 0.2, 0.3])
    network['W2'] = np.array([[0.1, 0.4, 0.7], [0.2, 0.5, 0.8], [0.3, 0.6, 0.9]])
    network['b2'] = np.array([0.1, 0.2, 0.3])

    return network

# パラメータの初期設定
network = init_network()
# ネットワークを検査
chk_network(network)
````
