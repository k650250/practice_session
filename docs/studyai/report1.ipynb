# レポート（応用数学）

## 第1章 : 線形代数
- スカラー・ベクトル・行列
    - スカラー
        - 階数$0$のテンソルに対応
        - いわゆる普通の数
            - 例.
        $$
        a=1
        $$
            - 大抵は実数$\mathbb{R}$
        - 四則演算可能
        - ベクトルに対する係数になれる
    - ベクトル
        - 階数$1$のテンソルに対応
        - 「大きさ」と「向き」を持ち、幾何学的座標系に於いて矢印（有向線分）で図示される
        - スカラーとセットで表示（成分表示）され、スカラーの要素を水平方向（横）に並べたものを「行ベクトル」、垂直方向（縦）に並べたものを「列ベクトル」という
        $$
        \left(
        \begin{array}{ccc}
            2 & 0 & -1
        \end{array}
        \right),\hspace{5pt}
        \left(
        \begin{array}{c}
            2 \\
            0 \\
            -1
        \end{array}
        \right)
        $$
        - ベクトルの文字表記は一般には$\rightarrow$記号を冠した小文字（例: $\vec{a}$）又は太字の小文字（例: $\mathbf{a}$）が使用される
        - 列ベクトル$\vec{a}$に紐付けられた$n$個のスカラーの要素を$\vec{a}$の「成分」といい、各成分固有の添字（インデックス）が割り当てられた列ベクトル
        $$
        \vec{a}=\left(
        \begin{array}{c}
            a_{1} \\
            a_{2} \\
            \vdots \\
            a_{n}
        \end{array}
        \right)
        $$
        を「$n$次元列ベクトル」という
            - $1$から$n$迄の範囲内の任意の成分を表わす添字には$i$を用いる
            $$
            1 \leqq i \leqq n \hspace{10pt} (i, n \in \mathbb{N})
            $$
            - $a_{i}$を$\vec{a}$の「第$i$成分」という
        - ソースコード例
        
        ```python
        import numpy as np
        
        # 列ベクトルのサイズ
        n = 3
        # 成分の値が一様乱数のn次元列ベクトル
        vec_a = np.random.rand(n)
        
        print(vec_a)
        ```
        
        - 演算
            - スカラーとベクトルの積（スカラー乗法）
        $$
        \begin{eqnarray}
            c\vec{a}&=&\left(
            \begin{array}{c}
                ca_{1} \\
                ca_{2} \\
                \vdots \\
                ca_{n}
            \end{array}
            \right) \\
            2\left(
            \begin{array}{c}
                2 \\
                -3 \\
                1
            \end{array}
            \right)
            &=&\left(
            \begin{array}{c}
                2\cdot 2 \\
                2\cdot (-3) \\
                2\cdot 1
            \end{array}
            \right)
            =\left(
            \begin{array}{c}
                4 \\
                -6 \\
                2
            \end{array}
            \right)
        \end{eqnarray}
        $$
    - 行列
        - 階数$2$のテンソルに対応
        - スカラーを表にしたもの
        $$
        \left(
        \begin{array}{cccc}
            1 & 3 & 0 & -1 \\
            2 & 5 & \frac{1}{2} & 6 \\
            -2 & 3 & -10 & \frac{1}{4} 
        \end{array}
        \right)
        $$
        - 行列の文字表記は一般には太字の大文字（例: $\mathbf{A}$）が使用される
        - 行列$\mathbf{A}$に紐付けられたスカラーの要素を$\mathbf{A}$の「成分」といい、各成分固有の添字が行と列の各軸毎に割り当てられた行列
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{m1} & a_{m2} & \ldots & a_{mn}
        \end{array}
        \right)
        $$
        を「$m$行$n$列の行列」又は「$m \times n$形行列」という
            
            - ソースコード例
            
            ```python
            import numpy as np
            
            # 行列の行数と列数
            m, n = 3, 4
            # 成分の値が一様乱数のmn形行列
            A = np.random.rand(m, n)
            
            print(A)
            ```
            
            - $1$から$m$迄の範囲内の任意の行ベクトル（タプル）を表わす添字には$i$を用いる
            $$
            1 \leqq i \leqq m \hspace{10pt} (i, m \in \mathbb{N})
            $$
            - 行ベクトル$a_{i}$を$\mathbf{A}$の「第$i$行」という
            - $1$から$n$迄の範囲内の任意の列ベクトル（カラム）を表わす添字には$j$を用いる
            $$
            1 \leqq j \leqq n \hspace{10pt} (j, n \in \mathbb{N})
            $$
            - 列ベクトル$a_{j}$を$\mathbf{A}$の「第$j$列」という
            - $a_{ij}$を$\mathbf{A}$の「$(i,j)$成分」という
            - $\mathbf{A}$の$(i,j)$成分を$(j,i)$成分の位置に移動して作った新たな行列
        $$
        \begin{eqnarray}
            \mathbf{A}^{\mathrm{T}}&=&\left(
            \begin{array}{cccc}
                a_{11} & a_{12} & \ldots & a_{1n} \\
                a_{21} & a_{22} & \ldots & a_{2n} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{m1} & a_{m2} & \ldots & a_{mn}
            \end{array}
            \right)^{\mathrm{T}}&=&\left(
            \begin{array}{cccc}
                a_{11} & a_{21} & \ldots & a_{m1} \\
                a_{12} & a_{22} & \ldots & a_{m2} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{1n} & a_{2n} & \ldots & a_{mn}
            \end{array}
            \right) \\
            &=&\left(
            \begin{array}{cccc}
                1 & 3 & 0 & -1 \\
                2 & 5 & \frac{1}{2} & 6 \\
                -2 & 3 & -10 & \frac{1}{4} 
            \end{array}
            \right)^{\mathrm{T}}
            &=&\left(
            \begin{array}{ccc}
                1 & 2 & -2 \\
                3 & 5 &  3 \\
                0 & \frac{1}{2} & -10 \\
                -1 & 6 & \frac{1}{4}
            \end{array}
            \right)
        \end{eqnarray}
        $$
            を$\mathbf{A}$の「転置行列」という
            - 行数と列数が等しい$(m=n)$行列
        $$
        \begin{eqnarray}
            \mathbf{A}&=&\left(
            \begin{array}{cccc}
                a_{11} & a_{12} & \ldots & a_{1n} \\
                a_{21} & a_{22} & \ldots & a_{2n} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{n1} & a_{n2} & \ldots & a_{nn}
            \end{array}
            \right) \\
            &=&\left(
            \begin{array}{ccc}
                1 & 3 & 0 \\
                2 & 5 & \frac{1}{2} \\
                -2 & 3 & -10
            \end{array}
            \right)
        \end{eqnarray}
        $$
            を「正方行列」といい、$n$行$n$列の正方行列を「$n$次正方行列」又は「$n$次行列」という
                
                - ソースコード例
                
                ```python
                import numpy as np
                
                # 正方行列のサイズ
                n = 3
                # 成分の値が一様乱数のn次正方行列
                A = np.random.rand(n, n)
                
                print(A)
                ```

            - $a_{11},a_{22},\ldots,a_{nn}$の成分を「対角成分」といい、対角成分を除く全ての成分の値を$0$とした正方行列
        $$
        \begin{eqnarray}
            \mathbf{C}&=&\left(
            \begin{array}{cccc}
                c_{1} & & & {\LARGE 0} \\
                & c_{2} & & \\
                & & \ddots & \\
                {\LARGE 0} & & & c_{n}
            \end{array}
            \right) \\
            &=&\left(
            \begin{array}{ccc}
                1 & 0 & 0 \\
                0 & 5 & 0 \\
                0 & 0 & -10
            \end{array}
            \right)
        \end{eqnarray}
        $$
            を「対角行列」という
                
                - ソースコード例
                
                ```python
                import numpy as np

                # 正方行列のサイズ
                n = 3
                # 成分の値が一様乱数のn次正方行列
                A = np.random.rand(n, n)
                # Aの対角成分で構成された対角行列
                C = np.diag(np.diag(v=A))

                print(C)
                ```
                
            - 全ての対角成分$c_{1},c_{2},\ldots,c_{n}$の値を$1$とした対角行列
        $$
        \begin{eqnarray}
            \mathbf{I}&=&\left(
            \begin{array}{cccc}
                1 & & & {\LARGE 0} \\
                & 1 & & \\
                & & \ddots & \\
                {\LARGE 0} & & & 1
            \end{array}
            \right) \\
            &=&\left(
            \begin{array}{ccc}
                1 & 0 & 0 \\
                0 & 1 & 0 \\
                0 & 0 & 1
            \end{array}
            \right)
        \end{eqnarray}
        $$
            を「単位行列」という
                
                - ソースコード例
                
                ```python
                import numpy as np
                
                # 正方行列のサイズ
                n = 3
                # 成分の値が一様乱数のn次正方行列
                A = np.random.rand(n, n)
                # Aと同じサイズの単位行列
                I = np.identity(n=A.shape[0])
                
                print(I)
                ```
        
        - 演算
            - 行列と列ベクトルの積
            <div style="overflow: auto hidden;">
            $$
            \begin{eqnarray}
                \mathbf{A}\vec{x}=
                \left(
                \begin{array}{cccc}
                    a_{11} & a_{12} & \ldots & a_{1n} \\
                    a_{21} & a_{22} & \ldots & a_{2n} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    a_{m1} & a_{m2} & \ldots & a_{mn}
                \end{array}
                \right)
                \left(
                \begin{array}{c}
                    x_{1} \\
                    x_{2} \\
                    \vdots \\
                    x_{n}
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{c}
                    \sum_{i=1}^{n}a_{1i}x_{i} \\
                    \sum_{i=1}^{n}a_{2i}x_{i} \\
                    \vdots \\
                    \sum_{i=1}^{n}a_{mi}x_{i} \\
                \end{array}
                \right) \\
                \left(
                \begin{array}{cc}
                    3 & 1 \\
                    2 & -3 \\
                    1 & 4
                \end{array}
                \right)
                \left(
                \begin{array}{c}
                    1 \\
                    3
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{c}
                    3 \cdot 1 + 1 \cdot 3 \\
                    2 \cdot 1 + (-3) \cdot 3 \\
                    1 \cdot 1 + 4 \cdot 3
                \end{array}
                \right) \\
                &=& 
                \left(
                \begin{array}{c}
                    6 \\
                    -7 \\
                    13
                \end{array}
                \right)
            \end{eqnarray}
            $$
            </div>
                
                - 行列$\mathbf{A}$の列数$n$と列ベクトル$\vec{x}$の成分数$n$が等しいときにのみ積が成立する
            - 行列同士の積（内積、ドット積）
            <div style="overflow: auto hidden;">
            $$
            \begin{eqnarray}
                \mathbf{A}\mathbf{B}=
                \left(
                \begin{array}{cccc}
                    a_{11} & a_{12} & \ldots & a_{1n} \\
                    a_{21} & a_{22} & \ldots & a_{2n} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    a_{m1} & a_{m2} & \ldots & a_{mn}
                \end{array}
                \right)
                \left(
                \begin{array}{cccc}
                    b_{11} & b_{12} & \ldots & b_{1n} \\
                    b_{21} & b_{22} & \ldots & b_{2n} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    b_{m1} & b_{m2} & \ldots & b_{mn}
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{cccc}
                    \sum_{i=1}^{n}a_{1i}b_{i1} & \sum_{i=1}^{n}a_{1i}b_{i2} & \ldots & \sum_{i=1}^{n}a_{1i}b_{in} \\
                    \sum_{i=1}^{n}a_{2i}b_{i1} & \sum_{i=1}^{n}a_{2i}b_{i2} & \ldots & \sum_{i=1}^{n}a_{2i}b_{in} \\
                    \vdots & \vdots & \ddots & \vdots \\
                    \sum_{i=1}^{n}a_{mi}b_{i1} & \sum_{i=1}^{n}a_{mi}b_{i2} & \ldots & \sum_{i=1}^{n}a_{mi}b_{in}
                \end{array}
                \right) \\
                \left(
                \begin{array}{cc}
                    3 & 1 \\
                    2 & -3 \\
                    1 & 4
                \end{array}
                \right)
                \left(
                \begin{array}{ccc}
                    1 & -2 & 0 \\
                    3 &  1 & 2
                \end{array}
                \right)
                &=& 
                \left(
                \begin{array}{ccc}
                    3 \cdot 1 + 1 \cdot 3    & 3 \cdot (-2) + 1 \cdot 1    & 3 \cdot 0 + 1 \cdot 2 \\
                    2 \cdot 1 + (-3) \cdot 3 & 2 \cdot (-2) + (-3) \cdot 1 & 2 \cdot 0 + (-3) \cdot 2 \\
                    1 \cdot 1 + 4 \cdot 3    & 1 \cdot (-2) + 4 \cdot 1    & 1 \cdot 0 + 4 \cdot 2
                \end{array}
                \right) \\
                &=& 
                \left(
                \begin{array}{ccc}
                     6 & -5 &  2 \\
                    -7 & -7 & -6 \\
                    13 &  2 &  8
                \end{array}
                \right)
            \end{eqnarray}
            $$
            </div>
                
                - 可換則（交換法則）は一般には成立しない
                - 行列$\mathbf{A}$の列数$n$と行列$\mathbf{B}$の行数$m$が等しいときにのみ積が成立する
                - 行列$\mathbf{A}$は行ベクトルを垂直方向に並べたもの、行列$\mathbf{B}$は列ベクトルを水平方向に並べたものとも考えられる
        - 主な用途
            - ベクトル等データの変換
                - データフロー図（DFD）でいう「データ処理」に相当
            - 連立方程式（連立一次方程式）を解く
                - $ax=b$の形で表したい
                - $n$本の一次方程式を持つ連立方程式
        $$
        \left\{
        \begin{array}{rrrrrrrrr}
            a_{11}x_{1} &+& a_{12}x_{2} &+& \ldots &+& a_{1n}x_{n} &=& b_{1} \\
            a_{21}x_{1} &+& a_{22}x_{2} &+& \ldots &+& a_{2n}x_{n} &=& b_{2} \\
            \vdots &\,& \vdots &\,& \ddots &\,& \vdots &\,& \vdots \\
            a_{n1}x_{1} &+& a_{n2}x_{2} &+& \ldots &+& a_{nn}x_{n} &=& b_{n}
        \end{array}
        \right.
        $$
                    を$x$の係数$a$を成分に持つ$n$次正方行列$\mathbf{A}$、$x$を成分に持つ$n$次元列ベクトル$\vec{x}$、$b$を成分に持つ$n$次元列ベクトル$\vec{b}$
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{n1} & a_{n2} & \ldots & a_{nn}
        \end{array}
        \right), 
        \vec{x}=\left(
        \begin{array}{c}
            x_{1} \\
            x_{2} \\
            \vdots \\
            x_{n}
        \end{array}
        \right), 
        \vec{b}=\left(
        \begin{array}{c}
            b_{1} \\
            b_{2} \\
            \vdots \\
            b_{n}
        \end{array}
        \right)
        $$
                    に当て嵌めたとき
        $$
        \mathbf{A}\vec{x}=\vec{b}
        $$
                    と表わすことができる。このとき、$\mathbf{A}$を「係数行列」という
                    - **連立方程式の行列の形は、連立方程式の係数を取り出したもの**といえる
                    - 例. 連立方程式
        $$
        \left\{
        \begin{array}{rrrrr}
            2x &-& 3y &=& 7 \\
            3x &+& 4y &=& 2
        \end{array}
        \right.
        $$
                    を
        $$
        \mathbf{A}=\left(
        \begin{array}{cc}
            2 & -3 \\
            3 & 4
        \end{array}
        \right),\hspace{5pt}
        \vec{x}=\left(
        \begin{array}{c}
            x \\
            y
        \end{array}
        \right),\hspace{5pt}
        \vec{b}=\left(
        \begin{array}{c}
            7 \\
            2
        \end{array}
        \right)
        $$
                    に当て嵌めたとき
        $$
        \begin{eqnarray}
            \mathbf{A}\vec{x}&=&\vec{b} \\
            \left(
            \begin{array}{cc}
                2 & -3 \\
                3 & 4
            \end{array}
            \right)
            \left(
            \begin{array}{c}
                x \\
                y
            \end{array}
            \right)
            &=&\left(
            \begin{array}{c}
                7 \\
                2
            \end{array}
            \right)
        \end{eqnarray}
        $$
                    と表わすことができる
                - 係数行列$\mathbf{A}$と、別の行列又はベクトル（例: $\vec{b}$）を水平方向に連結（concatenate）して作った新たな行列
        $$
        \left(
        \begin{array}{c|c}
            \mathbf{A} & \vec{b}
        \end{array}
        \right)
        =\left(
        \begin{array}{cc|c}
            2 & -3 & 7 \\
            3 & 4 & 2
        \end{array}
        \right)
        $$
                を「拡大係数行列」という
                    - ソースコード例
                    
                    ```python
                    import numpy as np
                    
                    # 係数行列
                    A = np.array([[2, -3],
                                    [3, 4]], dtype=int)
                    # 係数行列と水平方向に連結する為の列ベクトル
                    vec_b = np.array([7, 2], dtype=int).reshape(-1, 1)
                    # 拡大係数行列
                    Ab = np.concatenate([A, vec_b], axis=1)
                    
                    print(A)
                    print(vec_b)
                    print(Ab)
                    ```

                - 連立方程式を解く為の3つの技法を「行基本変形」という
                    - $i$行目を$c$倍する
        $$
        \mathbf{Q}_{i,c}=\left(
        \begin{array}{ccccccc}
            1 & & & & & & {\LARGE 0} \\
            & \ddots & & & & & \\
            & & 1 & & & & \\
            & & & c & & & \\
            & & & & 1 & & \\
            & & & &  & \ddots & \\
            {\LARGE 0} & & & & & & 1
        \end{array}
        \right)
        $$
                    - $s$行目に$t$行目の$c$倍を加える
        $$
        \mathbf{R}_{s,t,c}=\left(
        \begin{array}{ccccccc}
            1 & & & & & & {\LARGE 0} \\
            & \ddots & & & & & \\
            & & 1 & & c & & \\
            & & & \ddots & & & \\
            & & & & 1 & & \\
            & & & & & \ddots & \\
            {\LARGE 0} & & & & & & 1
        \end{array}
        \right)
        $$
                    - $p$行目と$q$行目を入れ替える
        $$
        \mathbf{P}_{p,q}=\left(
        \begin{array}{ccccccc}
            1 & & & & & & {\LARGE 0} \\
            & \ddots & & & & & \\
            & & 0 & & 1 & & \\
            & & & \ddots & & & \\
            & & 1 & & 0 & & \\
            & & & & & \ddots & \\
            {\LARGE 0} & & & & & & 1
        \end{array}
        \right)
        $$
                - 拡大係数行列$(\mathbf{A}\mid\vec{b})$に対して行基本変形を適用し、連立方程式を解く
                    1. $2$行目に$1$行目の$-1$倍を加える
        $$
        \underline{\left(
        \begin{array}{cc}
            1 & 0 \\
            -1 & 1
        \end{array}
        \right)^{(1)}}
        \left(
        \begin{array}{cc|c}
            2 & -3 & 7 \\
            3 & 4 & 2
        \end{array}
        \right)
        =\left(
        \begin{array}{cc|c}
            2 & -3 & 7 \\
            1 & 7 & -5
        \end{array}
        \right)
        $$
                    1. $1$行目と$2$行目を入れ替える
        $$
        \underline{\left(
        \begin{array}{cc}
            0 & 1 \\
            1 & 0
        \end{array}
        \right)^{(2)}}
        \left(
        \begin{array}{cc|c}
            2 & -3 & 7 \\
            1 & 7 & -5
        \end{array}
        \right)
        =\left(
        \begin{array}{cc|c}
            1 & 7 & -5 \\
            2 & -3 & 7
        \end{array}
        \right)
        $$
                    1. $2$行目に$1$行目の$-2$倍を加える
        $$
        \underline{\left(
        \begin{array}{cc}
            1 & 0 \\
            -2 & 1
        \end{array}
        \right)^{(3)}}
        \left(
        \begin{array}{cc|c}
            1 & 7 & -5 \\
            2 & -3 & 7
        \end{array}
        \right)
        =\left(
        \begin{array}{cc|c}
            1 & 7 & -5 \\
            0 & -17 & 17
        \end{array}
        \right)
        $$
                    1. $2$行目を$-\frac{1}{17}$倍する
        $$
        \underline{\left(
        \begin{array}{cc}
            1 & 0 \\
            0 & -\frac{1}{17}
        \end{array}
        \right)^{(4)}}
        \left(
        \begin{array}{cc|c}
            1 & 7 & -5 \\
            0 & -17 & 17
        \end{array}
        \right)
        =\left(
        \begin{array}{cc|c}
            1 & 7 & -5 \\
            0 & 1 & -1
        \end{array}
        \right)
        $$
                    1. $1$行目に$2$行目の$-7$倍を加える
        $$
        \underline{\left(
        \begin{array}{cc}
            1 & -7 \\
            0 & 1
        \end{array}

        - 逆行列
            - 或る2つの$n$次正方行列$\mathbf{A},\mathbf{A}^{-1}$と単位行列$\mathbf{I}$があり、次の式
            $$
            \mathbf{A}\mathbf{A}^{-1}=\mathbf{A}^{-1}\mathbf{A}=\mathbf{I}
            $$
            を満たすとき、$\mathbf{A}^{-1}$を$\mathbf{A}$の逆行列という
            - 或る$n$次正方行列$\mathbf{A}$とその転置行列$\mathbf{A}^{\mathrm{T}}$と単位行列$\mathbf{I}$があり、次の式
            $$
            \mathbf{A}\mathbf{A}^{\mathrm{T}}
            =\mathbf{A}^{\mathrm{T}}\mathbf{A}=\mathbf{I}
            $$
            を満たすとき、$\mathbf{A}$を「直交行列」という
                - $\mathbf{A}$が直交行列のとき、$\mathbf{A}$の転置行列$\mathbf{A}^{\mathrm{T}}$は$\mathbf{A}$の逆行列$\mathbf{A}^{-1}$と等価になる
            $$
            \mathbf{A}^{\mathrm{T}}=\mathbf{A}^{-1}
            $$

        - 行列式
            - 定義<br />
            或る$n$次正方行列
            $$
            \mathbf{A}=\left(
            \begin{array}{cccc}
                a_{11} & a_{12} & \ldots & a_{1n} \\
                a_{21} & a_{22} & \ldots & a_{2n} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{n1} & a_{n2} & \ldots & a_{nn}
            \end{array}
            \right)
            $$
            に対し、その成分によって定義される式
            $$
            \sum\varepsilon(p_{1}p_{2}\ldots p_{n})a_{1p_{1}}a_{2p_{2}}\ldots a_{np_{n}}
            $$
            を$n$次の$\mathbf{A}$の行列式といい、
            $$
            \left|
            \begin{array}{cccc}
                a_{11} & a_{12} & \ldots & a_{1n} \\
                a_{21} & a_{22} & \ldots & a_{2n} \\
                \vdots & \vdots & \ddots & \vdots \\
                a_{n1} & a_{n2} & \ldots & a_{nn}
            \end{array}
            \right|
            $$
            で表し、略して$|\mathbf{A}|$、$\mathrm{det}\mathbf{A}$とも書く
            - 特徴
                - 同じ行ベクトルが含まれていると行列式は$0$
                $$
                \left|
                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots  \\
                    \vec{w} \\
                    \vdots  \\
                    \vec{w} \\
                    \vdots  \\
                    \vec{v}_{n}
                \end{array}
                \right|
                =0
                $$
                - 1つのベクトルが$\lambda$倍されると行列式は$\lambda$倍される（線形性）
                $$
                \left|
                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots \\
                    \lambda\vec{v}_{i} \\
                    \vdots \\
                    \vec{v}_{n}
                \end{array}
                \right|
                =\lambda\left|
                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots \\
                    \vec{v}_{i} \\
                    \vdots \\
                    \vec{v}_{n}
                \end{array}
                \right|
                $$

                $$
                \left|
                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots \\
                    \vec{v}_{i}+\vec{w} \\
                    \vdots \\
                    \vec{v}_{n}
                \end{array}
                \right|

                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots \\
                    \vec{v}_{i} \\
                    \vdots \\
                    \vec{v}_{n}
                \end{array}
                \right|
                +\left|
                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots \\
                    \vec{w} \\
                    \vdots \\
                    \vec{v}_{n}
                \end{array}
                \right|
                $$

                    \vdots \\
                    \vec{v}_{s} \\
                    \vdots \\
                    \vec{v}_{r} \\
                    \vdots \\
                    \vec{v}_{n}
                \end{array}
                \right|

                \begin{array}{c}
                    \vec{v}_{1} \\
                    \vdots \\
                    \vec{v}_{r} \\
                    \vdots \\
                    \vec{v}_{s} \\
                    \vdots \\

                \right|

                    - 証明<br />
                    順列の異なる要素を交換する操作を「互換」といい、偶数回の互換で作れる順列を「偶順列（even permutation）」、奇数回の互換で作れる順列を「奇順列」という。順列に対し、その偶奇に従って
                $$
                    % !TEX encoding = UTF-8 Unicode
                    \varepsilon(i, j, \ldots, k) =
                    \begin{cases}
                        +1 & ((i, j, \ldots, k)\text{ is the even permutation}) \\
                        -1 & ((i, j, \ldots, k)\text{が奇順列})
                    \end{cases}
                $$
                    を定義し「順列$(i, j, \ldots, k)$の符号」という。
                    このように、順列の符号は互換によって変わるので
                $$
                    \varepsilon(i, \ldots, j, \ldots, k, \ldots, l)=-\varepsilon(i, \ldots, k, \ldots, j, \ldots, l)
                $$
                    を満たす。よって、$n$次の行列式
                $$
                    \mathrm{det}\mathbf{A}=\left|
                    \begin{array}{cccc}
                        a_{11} & a_{12} & \ldots & a_{1n} \\
                        a_{21} & a_{22} & \ldots & a_{2n} \\
                        \vdots & \vdots & \ddots & \vdots \\
                        a_{n1} & a_{n2} & \ldots & a_{nn}
                    \end{array}
                    \right|
                $$
                    に対し
                $$
                    \begin{eqnarray}
                        \text{左辺} &=& \sum\varepsilon(i, \ldots, j, \ldots, k, \ldots, l)a_{1i}\ldots a_{rj}\ldots a_{sk} \ldots a_{nl} \\
                        &=& \sum\varepsilon(i, \ldots, j, \ldots, k, \ldots, l)a_{1i}\ldots a_{sk}\ldots a_{rj} \ldots a_{nl} \\
                        &=& \sum\{-\varepsilon(i, \ldots, k, \ldots, j, \ldots, l)a_{1i}\ldots a_{sk}\ldots a_{rj} \ldots a_{nl}\} \\
                        &=& \text{右辺}
                    \end{eqnarray}
                $$
                    を満たすので、行列式の行を交換すると符号が変わる

- 固有値・固有ベクトル、特異値・特異ベクトル
    - 定義<br />
    或る$n$次正方行列
    $$
    \mathbf{A}=\left(
    \begin{array}{cccc}
        a_{11} & a_{12} & \ldots & a_{1n} \\
        a_{21} & a_{22} & \ldots & a_{2n} \\
        \vdots & \vdots & \ddots & \vdots \\
        a_{n1} & a_{n2} & \ldots & a_{nn}
    \end{array}
    \right)
    $$
    に対し、$0$でない特殊なベクトル$\vec{x}$とスカラーの定数$\lambda$が存在して
    $$
    \mathbf{A}\vec{x}=\lambda\vec{x}
    $$
    を満たすとき、$\lambda$を$\mathbf{A}$の「固有値」、$\vec{x}$を$\lambda$に対する$\mathbf{A}$の「固有ベクトル」という
        - 例.
        $$
        \mathbf{A}=\left(
        \begin{array}{cc}
            4 & 2 \\
            -3 & -1
        \end{array}
        \right),\hspace{5pt}
        \vec{x}=\left(
        \begin{array}{c}
            2 \\
            -3
        \end{array}
        \right)
        $$
        のとき
        $$
        \begin{eqnarray}
            \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
            \left(
            \begin{array}{cc}
                4 & 2 \\
                -3 & -1
            \end{array}
            \right)
            \left(
            \begin{array}{c}
                2 \\
                -3
            \end{array}
            \right)
            &=& \lambda \left(
            \begin{array}{c}
                2 \\
                -3
            \end{array}
            \right)
        \end{eqnarray}
        $$
        より、$\lambda=1$のとき、$\left(\begin{array}{c}2 \\ -3\end{array}\right)$は、固有値$1$に対する$\left(\begin{array}{cc} 4 & 2 \\ -3 & -1 \end{array}\right)$の固有ベクトル（の定数倍）である
    - 固有値・固有ベクトルの求め方
    $$
    \begin{eqnarray}
        \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
        (\mathbf{A}-\lambda I)\vec{x}&=&\vec{0}
    \end{eqnarray}
    $$
    $\mathbf{A}-\lambda I$の逆行列$(\mathbf{A}-\lambda I)^{-1}$が存在しないこと、かつ$\vec{x} \neq 0$より
    $$
    |\mathbf{A}-\lambda I|=0
    $$
    を満たすとき、この方程式を「固有方程式」といい、その左辺を「固有多項式」という
        - 例.
    $$
    \mathbf{A}=\left(
    \begin{array}{cc}
        2 & 1 \\
        3 & 4
    \end{array}
    \right)
    $$
        のとき、固有方程式より
    $$
    \begin{eqnarray}
        |\mathbf{A}-\lambda I|&=&0 \\
        \left|
        \begin{array}{cc}
            2-\lambda & 1 \\
            3 & 4-\lambda
        \end{array}
        \right|&=&0 \\
        (2-\lambda)(4-\lambda)-1\cdot3&=&0 \\
        \lambda^{2}-6\lambda+5&=&0 \\
        (\lambda-1)(\lambda-5)&=&0 \\
        \lambda&=&5\,\mathrm{or}\,1
    \end{eqnarray}
    $$
        したがって、$\mathbf{A}$の固有値$\lambda$は$5$と$1$の2つである<br /><br />$\lambda=5$のとき
    $$
    \begin{eqnarray}
        \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
        \left(
        \begin{array}{cc}
            2 & 1 \\
            3 & 4
        \end{array}
        \right)
        \left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        &=&5\left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        \; \Longleftrightarrow \;
        \left\{
        \begin{array}{rrrrr}
            2x_{1} &+& x_{2} &=& 5x_{1} \\
            3x_{1} &+& 4x_{2} &=& 5x_{2}
        \end{array}
        \right.
    \end{eqnarray}
    $$
        よって
    $$
    3x_{1} = x_{2}
    $$
        したがって、固有値$\lambda_{1}=5$に対する$\mathbf{A}$の固有ベクトル$\vec{v}_{1}$は$\left(\begin{array}{c} 1 \\ 3 \end{array}\right)$の定数倍である<br />$\lambda=1$のとき
    $$
    \begin{eqnarray}
        \mathbf{A}\vec{x}&=&\lambda\vec{x} \\
        \left(
        \begin{array}{cc}
            2 & 1 \\
            3 & 4
        \end{array}
        \right)
        \left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        &=&\left(
        \begin{array}{c}
            x_{1} \\
            x_{2}
        \end{array}
        \right)
        \; \Longleftrightarrow \;
        \left\{
        \begin{array}{rrrrr}
            2x_{1} &+& x_{2} &=& x_{1} \\
            3x_{1} &+& 4x_{2} &=& x_{2}
        \end{array}
        \right.
    \end{eqnarray}
    \\
    $$
        よって
    $$
    -x_{1} = x_{2}
    $$
        したがって、固有値$\lambda_{2}=1$に対する$\mathbf{A}$の固有ベクトル$\vec{v}_{2}$は$\left(\begin{array}{c} 1 \\ -1 \end{array}\right)$の定数倍である
    - 固有値分解（スペクトル分解）
        - 或る$n$次正方行列
        $$
        \mathbf{A}=\left(
        \begin{array}{cccc}
            a_{11} & a_{12} & \ldots & a_{1n} \\
            a_{21} & a_{22} & \ldots & a_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            a_{n1} & a_{n2} & \ldots & a_{nn}
        \end{array}
        \right)
        $$
        が固有値$\lambda_{1},\lambda_{2},\ldots,\lambda_{n}$と固有ベクトル$\vec{v}_{1},\vec{v}_{2},\ldots,\vec{v}_{n}$を持つとき、各固有値の並びを対角成分とした対角行列
        $$
        \mathbf{\Lambda}=\left(
        \begin{array}{cccc}
            \lambda_{1} & & & {\LARGE 0} \\
            & \lambda_{2} & & \\
            & & \ddots & \\
            {\LARGE 0} & & & \lambda_{n}
        \end{array}
        \right)
        $$
        と、それに対する固有ベクトルを並べた行列
        $$
        \mathbf{V}=\left(
        \begin{array}{cccc}
            \vec{v}_{1} & \vec{v}_{2} & \ldots & \vec{v}_{n}
        \end{array}
        \right)
        $$
        を用意したとき、それらは
        $$
        \begin{eqnarray}
            \mathbf{A}\mathbf{V}&=&\mathbf{V}\mathbf{\Lambda} \\
            \mathbf{A}&=&\mathbf{V}\mathbf{\Lambda}\mathbf{V}^{-1}
        \end{eqnarray}
        $$
        を満たす
            - $\mathbf{\Lambda}$の成分$\lambda_{i}$の値の順序は基本的に降順
            - $\mathbf{V}$は定数倍なので一つに定まらない
                - 固有値分解に於いて、より重要なのは$\mathbf{\Lambda}$の方
            - 例
            $$
            \mathbf{A}=\left(
            \begin{array}{cc}
                2 & 1 \\
                3 & 4
            \end{array}
            \right),\hspace{5pt}
            \lambda_{1}=5,\hspace{5pt} \lambda_{2}=1,\hspace{5pt}
            \vec{v}_{1}=\left(\begin{array}{c} 1 \\ 3 \end{array}\right),\hspace{5pt}
            \vec{v}_{2}=\left(\begin{array}{c} 1 \\ -1 \end{array}\right)
            $$
            のとき
        $$
        \begin{eqnarray}
            \mathbf{A}&=&\mathbf{V}\mathbf{\Lambda}\mathbf{V}^{-1} \\
            \mathbf{A}
            &=&
            \left(
            \begin{array}{cc}
                \vec{v}_{1} &  \vec{v}_{2} \\
                & 
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                \lambda_{1} & 0 \\
                0           & \lambda_{2}
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                \vec{v}_{1} &  \vec{v}_{2} \\
                & 
            \end{array}
            \right)^{-1} \\
            \left(
            \begin{array}{cc}
                2 & 1 \\
                3 & 4
            \end{array}
            \right)
            &=&
            \left(
            \begin{array}{cc}
                1 &  1 \\
                3 & -1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                5 & 0 \\
                0 & 1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                1 &  1 \\
                3 & -1
            \end{array}
            \right)^{-1} \\
            \left(
            \begin{array}{cc}
                2 & 1 \\
                3 & 4
            \end{array}
            \right)
            &=&
            \left(
            \begin{array}{cc}
                1 &  1 \\
                3 & -1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                5 & 0 \\
                0 & 1
            \end{array}
            \right)
            \left(
            \begin{array}{cc}
                \frac{1}{4} &  \frac{1}{4} \\
                \frac{3}{4} & -\frac{1}{4}
            \end{array}
            \right)
        \end{eqnarray}
        $$
        - 分解することのメリット
            - 行列の累乗の計算が容易になる
            - 行列自体の特徴がわかる
    - 特異値分解（SVD）
        - 固有値分解が$n$次正方行列のみを対象とするのに対し、特異値分解は$m \times n$形行列
        $$
        \mathbf{M}=\left(
        \begin{array}{cccc}
            m_{11} & m_{12} & \ldots & m_{1n} \\
            m_{21} & m_{22} & \ldots & m_{2n} \\
            \vdots & \vdots & \ddots & \vdots \\
            m_{m1} & m_{m2} & \ldots & m_{mn}
        \end{array}
        \right)
        $$
        を対象とする
        - 条件
        $$
        \begin{eqnarray}
            \mathbf{M}\vec{v}&=&\sigma\vec{u} \\
            \mathbf{M}^{\mathrm{T}}\vec{u}&=&\sigma\vec{v}
        \end{eqnarray}
        $$
        を満たす特殊な単位ベクトル$\vec{v},\vec{u}$があるとき、特異値分解が可能
        - 特異値の求め方
        <div style="overflow: auto hidden;">
        $$
        \begin{array}{rcl|rcl|rcccl}
            \mathbf{M}\mathbf{V}&=&\mathbf{U}\mathbf{S} & \mathbf{M}^{\mathrm{T}}\mathbf{U}&=&\mathbf{V}\mathbf{S}^{\mathrm{T}} &
            \mathbf{V}\mathbf{V}^{\mathrm{T}}
            &=&\mathbf{V}^{\mathrm{T}}\mathbf{V}&=&\mathbf{I} \\
            \mathbf{M}&=&\mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}} & \mathbf{M}^{\mathrm{T}}&=&\mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}} &
            \mathbf{U}\mathbf{U}^{\mathrm{T}}
            &=&\mathbf{U}^{\mathrm{T}}\mathbf{U}&=&\mathbf{I} \\
        \end{array}
        $$
        </div>
        $\mathbf{V}$と$\mathbf{U}$は直交行列、$\mathbf{I}$は単位行列であり、これらの積は
        <div style="overflow: auto hidden;">
        $$
        \begin{array}{rcl|rcl}
            \mathbf{M}\mathbf{M}^{\mathrm{T}}
            &=&\mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}}
            \mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}} &
            \mathbf{M}^{\mathrm{T}}\mathbf{M}
            &=&\mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}}
            \mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}} \\
            &=&\mathbf{U}\mathbf{S}\mathbf{I}
            \mathbf{S}^{\mathrm{T}}\mathbf{U}^{\mathrm{T}} &
            &=&\mathbf{V}\mathbf{S}^{\mathrm{T}}\mathbf{I}
            \mathbf{S}\mathbf{V}^{\mathrm{T}} \\
            &=&\underline{\mathbf{U}\mathbf{S}\mathbf{S}^{\mathrm{T}}}
            \mathbf{U}^{\mathrm{T}} &
            &=&\mathbf{V}
            \underline{\mathbf{S}^{\mathrm{T}}\mathbf{S}\mathbf{V}^{\mathrm{T}}}
        \end{array}
        $$
        </div>となる
        
            - $\mathbf{M}\mathbf{M}^{\mathrm{T}}$は$m$次正方行列、$\mathbf{M}^{\mathrm{T}}\mathbf{M}$は$n$次正方行列となる
            - $\mathbf{M}\mathbf{M}^{\mathrm{T}}$を固有値分解すれば対角行列$\mathbf{S}\mathbf{S}^{\mathrm{T}}$の対角成分から特異値の2乗$\sigma^{2}$、$m$次正方行列$\mathbf{U}$から$m$次元列ベクトルの左特異ベクトル$\vec{u}$が求められる
            - $\mathbf{M}^{\mathrm{T}}\mathbf{M}$を固有値分解すれば対角行列$\mathbf{S}^{\mathrm{T}}\mathbf{S}$の対角成分から特異値の2乗$\sigma^{2}$、$n$次正方行列$\mathbf{V}^{\mathrm{T}}$から$n$次元列ベクトルの右特異ベクトル$\vec{v}$が求められる
            - 左右両特異$\vec{u},\vec{v}$ベクトルは単位ベクトルから作らなければならない
            - 求めた特異値の2乗の平方根$\sqrt{\sigma^{2}}=\pm{\sigma}$を用いて、それを対角成分とした対角行列$\mathbf{S}$を求める
            - 最終的に次の式が求まる
            $$
            \mathbf{M}=\mathbf{U}\mathbf{S}\mathbf{V}^{\mathrm{T}}
            $$
    - 主な用途
        - データ量を減らす
            - 主成分分析による次元削減

## 第2章 : 確率・統計
- 集合
    - ものの集まり<br />
    - 或る集合を数学的には
    $$
    S=\{a,b,c,d,e,f,g\}
    $$
    と表現する
        - $a$は集合$S$を構成する「要素」（元）の一つ
        $$
        a \in S
        $$
        - $h$は集合$S$を構成する要素ではない
        $$
        h \notin S
        $$
        - 内部に入れ子の集合$M=\{c,d,g\}$がある
        $$
        M \subset S
        $$
    - 確率・統計でいう「事象」は集合として取り扱うことができる
    - 全領域を網羅する集合$U$を「全集合」という
    - 和集合
        - 集合$A$と集合$B$を合わせた領域の集合
            $$
            A \cup B
            $$
    - 共通部分
        - 集合$A$と集合$B$に共通する領域の集合（積集合ではない）
            $$
            A \cap B
            $$
    - 絶対補
        - 集合$A$の領域を除く全集合
            $$
            U \backslash A = \overline{A}
            $$
    - 相対補
        - 集合$A$の領域を除く集合$B$の領域の集合
            $$
            B \backslash A = \overline{A}
            $$
    - 可換則
        - 集合$A \cap B$を集合$B \cap A$のように項を交換しても等価
            $$
            A \cap B=B \cap A
            $$
    
- 確率
    - 定義<br />或る事象$A$が起こる数に全事象の数を除した連続値
    $$
    P(A)=\frac{n(A)}{n(U)}=\frac{\text{事象Aが起こる数}}{\text{全事象の数}}
    $$
    を事象$A$が起こる「確率」という
        - $P(\cdot)$の値は$0$から$1$迄の連続値
        $$
        0 \leqq P(\cdot) \leqq 1 \hspace{10pt} (P(\cdot) \in \mathbb{R})
        $$
            - 百分率で表記することもある
        - $U$は「全事象」、$n(U)$は「全事象の数」を表わす
            - $n(U)$は$0$より大きい離散値又は連続値
            $$
            n(U) > 0
            $$
        - $A$は$U$に含まれる「事象」の一つ、$n(A)$は「事象$A$が起こる数」を表わす
        $$
        A \subset U
        $$
            - $n(A)$は$0$から$n(U)$迄の範囲内の離散値又は連続値
            $$
            0 \leqq n(A) \leqq n(U)
            $$
    - 全事象$U$の確率$P(U)$は$1$と等価
    $$
    \begin{eqnarray}
        P(U)&=&\frac{\text{全事象の数}}{\text{全事象の数}} \\
        &=&\frac{n(U)}{n(U)}
        &=&1
    \end{eqnarray}
    $$
    - $P(A)$の反対の確率は$1-P(A)$と等価
    $$
    \begin{eqnarray}
        P(\overline{A})&=&\frac{\text{事象Aが起こらない数}}{\text{全事象の数}} \\
        &=&\frac{\text{全事象の数}-\text{事象Aが起こる数}}{\text{全事象の数}} \\
        &=&\frac{n(U)-n(A)}{n(U)}
        &=&\frac{n(U)}{n(U)}&-&\frac{n(A)}{n(U)} \\
        &&&=&P(U)&-&P(A) \\
        &&&=&1&-&P(A)
    \end{eqnarray}
    $$
    - 同時確率
        - 定義:<br />或る事象$A$と、また別の或る事象$B$が共通の事象として同時に起こる確率
        $$
        P(A \cap B)=P(\mbox{事象A} \cap \mbox{事象B})
        $$        
        を「同時確率」という
        - 暗黙の前提である全事象$U$を式に表わすと
        $$
        P(A \cap B)=\frac{P(A \cap B)}{P(U)}
        $$
        となる
            - $P(U)=1$である
            - 一般には$U$が明記されることはない
        - $A$と$B$が互いに因果関係のない独立な事象のとき、互いの事象の確率同士の積
        $$
        P(A \cap B)=P(A)P(B)
        $$
        に変形できる
        - $A \cup B$の中にその共通部分$A \cap B$が存在するとき、$A \cup B$の確率は互いの事象の確率同士の和に対し同時確率を引いたものとなる
        $$
        P(A \cup B)=P(A)+P(B)-P(A \cap B)
        $$
            - 共通部分$A \cap B$が存在しないとき、同時確率は$0$になる
        - $A \cap B$は可換則により
        $$
        A \cap B=B \cap A
        $$
        なので
        $$
        P(A \cap B)=P(B \cap A)
        $$
        を満たす
    - 条件付き確率
        - 定義:<br />或る条件となる事象$B$が与えられた下で、また別の或る事象$A$起こる確率
        $$
        P(A \mid B)=P(\mbox{事象A} \mid \mbox{条件となる事象B})
        $$
        を「条件付き確率」という
        - 暗黙の前提である全事象$U$を考えたとき、事象$A$が起こる確率$P(A)$と、条件となる事象$U$が与えられた下で、事象$A$が起こる確率$P(A \mid U)$は等価である
        $$
        P(A)=P(A \mid U)
        $$
            - 一般には$U$が明記されることはない
        - 同時確率$P(A \cap B)$は、条件付き確率を持つ式に分解できる
        $$
        P(A \cap B)=P(A)P(B \mid A)
        $$
        - $A \cap B$は可換則により
        $$
        A \cap B=B \cap A
        $$
        なので、$P(A)>0$かつ$P(B)>0$のとき
        $$
        \begin{eqnarray}
            P(A \cap B)&=&P(B \cap A)\tag{1} \\
            P(A)P(B \mid A)&=&P(B)P(A \mid B)\tag{2} \\
            P(B)P(A \mid B)&=&P(A \cap B)\tag{3} \\
            P(B \mid A)&=&\frac{P(A \cap B)}{P(A)}\tag{4} \\
            P(A \mid B)&=&\frac{P(A \cap B)}{P(B)}\tag{5}
        \end{eqnarray}
        $$
        を満たす
            - $(4)$の式は、一般に「ベイズ則」（ベイズの定理）の代表式に挙げられる
        - 項が同じとき、条件付き確率は同時確率以上の数になる
        $$
        \begin{eqnarray}
            P(A \mid B) & \geqq & P(A \cap B) \\
            \frac{P(A \cap B)}{P(B)} & \geqq & \frac{P(A \cap B)}{P(U)}
        \end{eqnarray}
        $$
    - 主客別分類
        - 頻度確率（客観確率）
            - 発生する頻度
            - 例: サイコロ、籤引き
        - ベイズ確率（主観確率）
            - 信念の度合い
            - 直感的には粗筋でレポートを書き進めて行って途中で説明が不足している感じたら遡って補足して行く
- 分散と共分散
    - 分散
    $$
    \begin{eqnarray}
        \mathrm{Var}(f)&=&\mathrm{E}\left\{\left(f_{(X-x)}-\mathrm{E}_{(f)}\right)^{2}\right\} \\
        &=&\mathrm{E}\left(f_{(X-x)}^{2}\right)-\left(\mathrm{E}_{(f)}\right)^{2}
    \end{eqnarray}
    $$
        - データの散らばり具合
        - データの各々の値が、期待値$\mathrm{E}$からどれだけ乖離しているか平均したもの
        - 負の値にならない
            - 絶対値をとるよりも2乗した方が計算コストが低い
    - 共分散
    $$
    \begin{eqnarray}
        \mathrm{Cov}(f,g)&=&\mathrm{E}\left\{\left(f_{(X-x)}-\mathrm{E}(f)\right)\left(g_{(Y-y)}-\mathrm{E}(g)\right)\right\} \\
        &=&\mathrm{E}(fg)-\mathrm{E}(f)\mathrm{E}(g)
    \end{eqnarray}
    $$
        - 2つのデータ系列の傾向の違い
            - 正の値をとれば似た傾向
            - 負の値をとれば逆の傾向
            - $0$をとれば関係性に乏しい
